# 학습이 잘 되는 모델

- 사용할 레이어
    - tf.keras.layers.BatchNormaliztion()
    - tf.keras.layers.Activation('swish')
- 데이터
    - 보스턴 집값 예측
    https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv

```python
# 라이브러리 사용

import pandas as pd
import tensorflow as tf
```

```python
# 데이터를 준비합니다.

url = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv"
boston = pd.read_csv(url)

print(boston.dtypes)
print(boston.shape)
print(boston.columns)
```

```python
# 종속변수, 독립변수

독립 = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]

종속 = boston[['medv']]

print(독립.shape, 종속.shape)
```

```python
# 모델의 구조를 생성합니다.

X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss="mse")
```

```python
# 모델을 학습(fit)합니다.

model.fit(독립, 종속, epochs=1000)
```

Epoch 1000/1000

여러번 학습을 시켜도 loss값이 만족스럽지 못합니다.

이럴경우 어떻게 해결할까요??

핵심은 bach normalizaiton입니다.

학습을 더 빠르고 성능을 높일 수 있습니다.

activation layer를 분리해서 다음과 같이 작성할 수 있습니다.

```python
# 2. 모델의 구조를 BatchNormalization layer를 사용하여 만든다.
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

수식을 동일하지만 분리해주는 이유는 batch normalization을 Dense, Activation Layer 사이에 넣어주는 것이 효과가 좋기 때문입니다. 

```python
# Batch normalization을 Dense, Activation layer 사이에 넣어준 모델로 학습을 시켜보겠습니다.

model.fit(독립, 종속, epochs=1000)
```

```python
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(파일경로)

# 원핫인코딩
아이리스 = pd.get_dummies(아이리스)

# 종속변수, 독립변수
독립 = 아이리스[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 아이리스[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
# 2. 모델의 구조를 만듭니다
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy',
              metrics='accuracy')
```

```python
# 3.데이터로 모델을 학습(FIT)합니다.
model.fit(독립, 종속, epochs=1000)
```

accuracy: 0.9800
