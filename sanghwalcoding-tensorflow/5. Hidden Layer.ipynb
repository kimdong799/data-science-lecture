입력부분을 Input Layer, 출력 부분을 Hidden Layer라고 합니다.

같은 표현을 가지고 중간 레이어를 바라보면 히든레이어에서 나타난 데이터를 추론해볼 수 있습니다. 5개의 컬럼을 가진 데이타라구요

마지막 히든레이어에서 결과를 출력하는 부분만 살펴보면 5개의 입력을 받고 하나의 출력을 만드는 모델로 볼 수 있습니다.

입력쪽에서 히든레이어를 만든 부분을 떼어서 관찰해보면 13개의 입력을 받고 5개의 출력을 만드는 모델이라고 할 수 있습니다.
