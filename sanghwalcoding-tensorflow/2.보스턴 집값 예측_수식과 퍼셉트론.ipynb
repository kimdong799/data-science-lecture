```python
# 라이브러리 사용

import pandas as pd
import tensorflow as tf
```

```python
url = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv"
boston = pd.read_csv(url)

print(boston.columns)
boston.head()
```

```python
# 독립변수, 종속변수 분리

x = boston[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio',
            'b','lstat']]

y = boston[['medv']]

print(x.shape, y.shape)
```

```python
# 모델의 구조를 생성합니다.

X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 데이터로 모델을 학습(FIT)합니다.

model.fit(x,y, epochs=1000)
```

```python
# 모델을 이용합니다.

model.predict(x[0:5]) # 슬라이싱을 활용하여 506개의 x변수 중 5개만을 사용해 검증합니다.
```

```python
# 모델의 수식 확인

model.get_weights()
```

- 집값 = -0.091169618 * x1 + 0.063605994 * x2 + ......-0.51774025 * x13 + 3.1532197

- 모델을 통해서 위의 수식을 컴퓨터가 스스로 학습한 것입니다.

--------

우리는 위의 과정을 거쳐 딥러닝 모델을 완성합니다.



보스턴 파일을 불러온 다음 13개의 컬럼을 독립변수로, 예측하고자 하는 값을 종속변수로 분류했습니다.

이후 shape, Dense(n) 에 각각 독립변수, 종속변수의 개수를 넣어줍니다.

모델을 이렇게 구성해주면 앞서 보았던 복잡한 수식을 알아서 처리해주는 모델이 완성됩니다.

모델이 어떻게 동작하는지를 알아보기 위해 #2의 코드로 알아보겠습니다.

보스턴 집값의 독립변수가 13개이기 때문에 입력 레이어에서 13이라는 숫자를 사용했습니다.

첫줄의 코드는 그림으로 표현하면 다음과 같습니다.

13개의 입력을 받는 입력층을 구성한다 할 수 있습니다.

종속변수가 1개이기에 Dense에서는 1을 입력했습니다.

두번째 줄을 그림으로 표현하면 1개의 출력을 만드는 출력층을 만든것이라고 할 수 있습니다.

좀 더 정확히 표현하면 13개의 입력으로부터 1개의 출력을 만드는 구조를 만든것입니다.

이를 수식으로 표현하면 다음과 같습니다.

Dense 레이어는 이 수식을 만드는 것이고 우리가 만든 모델이 학습과정에서 입력되는 데이터를 보고 이 수식의 W와 b를 찾는 것입니다.

저희가 만든 모델은 뉴런 하나로 이루어져있습니다.

뉴런은 실제 두뇌세포의 이름이고 인공신경망에서 이런 뉴런의 역할을 하는 녀석이 바로 우리가 만든 모형과 수식입니다.

이 모형에는 퍼셉트론이라는 이름이 붙어있습니다. 각 w들을 부르는 이름은 가중치(weight) b는 편향(bias)라고 합니다.



데이터가 독립변수 12개 종속변수 2개라면 코드를 다음처럼 작성할 수 있습니다.

하나의 결과를 만드는 데에는 수식 하나가 필요합니다.

지금은 종속변수가 두개이니 하나의 결과가 더 필요합니다.

즉 결과와 수식이 2개가 필요한것입니다.

이 모양은 퍼셉트론 두개가 병렬로 연결된 모습이고 가중치의 개수는 첫번째 수식에서 w 12개 b 1개, 두번째 수식에서 w 12개, b 1개 총 26개의 값을 찾아야 할 것입니다.
