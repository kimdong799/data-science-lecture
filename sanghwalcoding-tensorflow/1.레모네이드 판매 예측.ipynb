```python
#  import

import pandas as pd
import tensorflow as tf
```

```python
#  

url = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv"
df = pd.read_csv(url)
df.head()
```

```python
# ,  

independent = df[[""]]
dependent = df[[""]]

print(independent.shape, dependent.shape)
```

```python
#  .

X = tf.keras.layers.Input(shape=[1]) # 1  
Y = tf.keras.layers.Dense(1)(X) #  
model = tf.keras.models.Model(X, Y)
model.compile(loss="mse")
```

```python
#  .
# verbose=0  X

model.fit(independent, dependent, epochs=1000)
```

```python
#  ()

model.predict(independent)
```
