모델은 매 학습마다 조금씩 학습하게 되는데 얼마나 정답에 가까이 맞추고 있는지를 평가하는 지표가 LOSS!

독립변수와 종속변수를 준비했고 모델을 만들었습니다.

독립변수를 모델에 넣어주면 모델은 예측결과를 만들어냅니다.

이를 평가하기 위해 실제 정답과 예측값을 비교합니다.

차이에 제곱 (예측-결과)^2 의 평균을 구하면 그 값이 LOSS입니다.

예측이 정답을 모두 맞추었다면 LOSS는 0이 될것입니다.

즉 0에 가까울수록 학습이 잘 된 모델이라고 할 수 있습니다.

epoch마다 loss가 0에 가까워지는지를 확인하는 것이 중요합니다.

# 실습

https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv

```python
# 라이브러리 import

import pandas as pd
import tensorflow as tf
```

```python
# 데이터 준비

url = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv"
df = pd.read_csv(url)
df.head()
```

```python
# 종속변수, 독립변수 분리

independent = df[["온도"]]
dependent = df[["판매량"]]

print(independent.shape, dependent.shape)
```

```python
# 모델을 생성합니다.

X = tf.keras.layers.Input(shape=[1]) # 1은 독립변수의 수
Y = tf.keras.layers.Dense(1)(X) # 종속변수의 수
model = tf.keras.models.Model(X, Y)
model.compile(loss="mse")
```

```python
# 모델을 학습합니다.
# verbose=0 은화면 출력X

model.fit(independent, dependent, epochs=1000)
```

```python
# 모델을 이용합니다(예측)

model.predict(independent)
```
