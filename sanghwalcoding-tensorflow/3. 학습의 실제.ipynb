딥러닝 데이터를 가지고 스스로 학습한다는 것은 매우 신기한 일입니다.

하지만 우리는 한가지를 알았습니다.

수식을 만들고 종속 변수의 값을 가장 근접하게 예측하도록 가중치를 찾는것이구나!

그렇다면 이 수식의 가중치는 어떻게 찾는 것일까요??

https://bit.ly/2DEBlPd

위의 주소로 들어가면 가중치 학습의 과정을 경험해 볼 수 있는 스프레드시트입니다.

최소 생성된 난수를 W, B에 입력합니다.

이 W, B값이 x변수(온도)에 다음의 수식으로 사용됩니다. 

예측(W * 온도 + b)

각각의 데이터에(row) 대해 (예측 - 판매량) 값을 구해 이를 제곱한 전체 데이터의 평균값을 구하면 MSE값이 나오게 됩니다. 이는 Mean Squared Error로 Loss값이 바로 이 MSE값입니다.

W값을 학습시키기 위해서는 미분을 통해 W가 어떤 방향으로 가야 Loss값을 줄일 수 있는지를 찾아야 합니다. 따라서 0.0001을 더해보고 Loss값이 커졌다면 W를 낮춰야 한다는 결론이 나오게 됩니다.

반대로 최초 생성된 난수에 0.0001을 더했는데 Loss값이 작아졌다면 W를 높여야 한다는 결론에 이르게 됩니다.

다만 무작정 키우는 것이 아니라 W에 0.0001을 더했을 때 나온 Loss값에서 기존의 Loss값을 뺀 값을 0.0001로 나눕니다.

W에서 위에서 나온 값을 빼고 0.0001을 곱하면 다음에 epoch에 사용될 W값이 정해지게 됩니다.

B값도 W와 같은 방식으로 구하게 됩니다.
