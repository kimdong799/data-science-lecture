실제 연산은 잠시 뒤로하고 Convolution layer를 이용한 딥러닝 코드를 먼저 확인해보겠습니다.

Convolution layer가 두 층 추가되었습니다.

Convolution layer에서 우리가 결정할 것은 필터셋을 몇개 사용할 것인지, 필터셋의 사이즈를 얼마로 할 것인지입니다.

3과 6은 필터셋의 갯수입니다.

3은 3개의 필터셋을 6은 6개의 필터셋을 사용하겠다는 의미입니다.

kernel_size는 필터셋의 크기를 의미합니다. 5, 5 사이즈의 필터 셋을 사용하기 위해서 5라고 입력합니다.

그리고 Convolution layer는 기존의 Dense layer와 같이 활성화 함수 activation 함수를 지정해주어야 합니다.

첫번째 Convolution layer는 필터셋이 3개이기 때문에 3장의 특징맵을 만듭니다. 다른 표현으로 3채널의 특징맵을 만든다고 할 수 있습니다.

2번째 Convolution layer는 필터셋을 6개 썼기 때문에 6채널의 특징맵을 만든다고 할 수 있습니다.

그리고 이렇게 만들어진 6채널의 특징맵 모두를 flatten 레이어에 의해서 픽셀단위로 한줄로 펼친 후 학습하게 됩니다.
