필터를 이해하는 것은 Convolution이 동작을 이해하는데 매우 큰 도움이 됩니다.

필터에 대해서 다음의 내용을 먼저 기억해주세요



필터셋은 3차원 형태로 된 가중치의 모음입니다.

Convolution 구조에서 컴퓨터가 학습하는 것은 바로 필터입니다.

필터셋 하나는 앞선 레이어의 결과인 특징맵 전체를 봅니다. 전체를 보고 필터셋 하나가 특징맵 하나를 만듭니다. 필터셋 하나가 특징맵 하나를 만드니 필터셋의 갯수만큼 특징맵을 만들게 되죠

이해를 위해 Convolution code를 가져왔습니다.

첫번째 레이어는 필터셋 3개, 두번째 레이어는 필터셋 6개입니다.

그리고 각 필터는 5,5 사이즈입니다.

3차원 형태로 된 가중치의 모음이라고 앞서 언급했습니다. 



이렇게 생각하면 안되기에 필터셋에 대한 설명을 제대로 하려고 하는 것입니다.

3차원의 필터셋이라는 것은 개별 필터셋 하나가 3차원 형태라는 것입니다.

즉, 5,5 사이즈의 필터가 여러 채널 존재하는 것입니다. 그리고 그 숫자는 앞선 레이어의 특징맵 전체를 본다라는 두번 째 설명과 관련이 있는데 전체를 보기 때문에 필터의 채널수 ? 에는 특징맵의 채널수와 같은 숫자가 들어가게 됩니다.

입력이 흑백이라고 생각해봅시다. 흑백의 경우 채널이 하나이니 필터셋의 형태는 5, 5, 1이 됩니다. 

입력이 만약 컬러 이미지라면 채널이 3이므로 필터셋의 형태도 5, 5, 3이 됩니다. 

또한 필터셋 전체의 모양은 다음과 같이 4차원 형태가 되는 것입니다. 필터셋이 3개니까 3, 5, 5, ? 가 되는 것이죠




제가 설명을 돕기 위해 이곳에서 필터셋이라는 표현을 사용하는데요 5, 5 의 필터가 채널수만큼 겹쳐진 형태이기 때문에 하지만 실제 이 분야에서는 이렇게 여러채널로 구성된 형태 그대로 필터라고 합니다.

방금 코드를 딥러닝 모델로 이해해보겠습니다.

입력 이미지가 있습니다. 28, 28, 1 의 흑백 이미지 입니다.

5, 5, 1 필터셋이 3개 준비되었고 필터셋 하나는 한장의 특징맵을 생성하므로 이 필터셋에 의하여 특징맵 세장 3채널의 특징맵이 생성됩니다. 

특징맵의 사이즈가 28에서 24로 줄었는데 그 이유는 필터의 사이즈가 5이기때문입니다, 사이즈에서 1을 뺀만큼 사이즈가 감소합니다.


두번째 레이어에서는 6장의 필터셋을 준비합니다. 필터셋의 채널수가 앞선 레이어의 결과인 특징맵 채널수와 같은 3으로 되어있습니다. 필터셋이 6개이니 6채널의 특징맵이 생성됩니다. 

특징맵 하나를 만들 때 앞의 특징맵 전체를 보고 새로운 특징맵 하나를 만드는 것입니다.

이전에 표의 특징을 학습했던 모델을 레이어와 노드로 표현한 그림입니다.

하나의 특징을 만들 때 앞선 레이어의 노드 전체를 사용하여 학습한다고 했습니다.



앞선 특징맵 전체를 보고 새로운 특징맵 하나를 만든다는 것은 이것과 같은 맥락입니다.

또한 이 노드의 84개의 특징은 컴퓨터가 가중치들을 학습해서 스스로 찾아낸 특징이고 그래서 딥러닝에는 특징 자동 추출기라는 별명이 붙어있다고 했었죠





Convolution 레이어에서는 필터들을 뜯어보면 가중치로 이루어져있고 컴퓨터는 특징맵을 만들어내는 컨볼루션 필터를 학습하는 것입니다. 

앞서 히든레이어를 추가하는 것은 지정한 노드 갯수만큼 컴퓨터에게 분류를 위한 가장 좋은 특징을 찾아달라고 요청하는 것이라고 했습니다.

따라서 6개의 필터를 사용한다는 것은 컴퓨터에게 이미지 분류를 위해 사용할 가장 좋은 특징맵 6개를 찾아달라고 요청하는 것입니다.
