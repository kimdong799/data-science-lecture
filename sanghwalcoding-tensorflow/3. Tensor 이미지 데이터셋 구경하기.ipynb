이번에는 우리가 앞으로 사용할 이미지들을 먼저 구경해보겠습니다.

앞서배운 차원의 관점으로 이미지 데이터를 바라보고 이해하는 연습을 해봅시다.

우리는 전체 과정에서 Mnis라는 손글씨 숫자 이미지 셋과 CIFAR10 이라고 하는 10가지 사물 분류를 위한 이미지 셋을 사용해 볼것입니다.

Mnis는 흑백, CIFAR10은 컬러 이미지입니다.

손글씨 이미지 하나를 살펴볼까요 5라는 숫자가 쓰여져 있네요

컴퓨터에게는 이 이미지가 숫자 5가 아닌 숫자들의 모음일 뿐입니다.

5라는 숫자가 보이시나요? 숫자들의 뭉치 사이에서 5라는 숫자가 보일거에요

숫자들의 집합이 2차원 형태 28 * 28 로  이루어져 있습니다.

각 숫자 하나는 점의 까맣고 하얀 정도를 나타냅니다.

검정 부분은 0 하얀 부분은 255에 가까운 수로 되어있는것이 보이시나요?

흑백의 정도를 표현하는 데이터셋입니다.

이 안의 숫자는 28*28= 784가 되겠네요

바로 784개의 숫자가 이 안에 있다는 것이죠

차원의 개념을 아는 우리는 이 이미지를 이렇게 말할 수 있습니다.

이 이미지 하나는 2차원 형태로 되어있고 784차원 공간의 한 점으로 표현 할 수 있다고 말이죠

만약 이 이미지가 6만 장 준비되어있다면 이미지셋 데이터의 형태는 6만, 28, 28이 되겠죠

CIFAR 10 이미지 하나를 가져와 보았습니다.

이 이미지 역시 컴퓨터에게는 이미지가 아니라 숫자들의 모음으로 인지됩니다.

가로로 32개, 세로 32개의 칸에 채워진 숫자들입니다.

컬러 이미지는 흑백과 다른점이 있는데요 32, 32 2차원 숫자 집합이 3개 존재합니ㅏㄷ.

각각 RGB를 나타내는 숫자들이죠

잠깐 앞서본 데이터를 가지고 설명하자면 흑백은 각 칸에 흑백에 대한 숫자가 하나씩 들어있는데

컬러는 그 칸에 숫자가 세개씩 들어있다고 생각하면 됩니다.

즉 32, 32 에 각 점이 세개의 값을 가지게 되는 것입니다.

0~ 255 사이의 값으로 하나는 R, 하나는 G, 하나는 B의 농도를 나타내는 것이죠

때문에 개별 CIFAR 10 이미지는 32, 32, 3의 모양이 되고 3차원 형태입니다.

32 * 32 * 3을 하면 3072인데요 CIFAR 10 이미지 하나는 3072개의 숫자가 들어있고 3072차원의 공간의 한점으로 표현된다고 말할 수 있겠습니다.

만약 이 이미지가 5만장 준비되어 있다면 이 이미지셋의 모양은 4차원인 5만, 32, 32, 3 이 됩니다.
