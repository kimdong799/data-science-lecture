이미지를 학습하는 가장 기본적인 딥러닝 모델로 CNN 이라는 새로운 딥러닝 구조도 배우게 됩니다.

지금 보시는 모델은 CNN 구조로 구현된 LeNet 이라는 이름의 모델입니다.

이미지 학습을 위한 첫번째 전체 코드를 구경해봅시다.

독립변수가 784개의 컬럼으로 구성되고 종속변수가 10개의 컬럼으로 구성된 표를 학습하는 것입니다.

0~9까지 10개의 범주형 숫자가 있으니 원핫인코딩을 하여 종속변수가 10개의 컬럼이 되는 것이 보입니다.

그런데 이미지를 학습하는데 표를 학습한다니? 좀 이상하지 않으신가요?

독립변수를 만드는 부분을 눈여겨보면 이 코드를 사용하는 것이 어렵지 않아 보일 것입니다.

여기 한장의 이미지 데이터가 있습니다. 28 * 28 2차원 형태입니다.

그리고 다음과 같은 표를 준비합니다. 컬럼이 784개로 되어있는 표입니다.

우리가 할 일은 준비한 이미지 데이터를 이 표에 넣는 것입니다.

기존의 60000, 28, 28 shape의 이미지 데이터를 784개로 구성된 표에 reshape 함수를 통해 집어넣을 수 있습니다.

데이터가 표로 바뀌면 우리는 아무런 두려움이 없습니다.

칼럼이 784개인 독립변수, 칼럼이 10개인 종속변수를 가지고 학습을 한다면 입력층의 노드는 784개 출력층은 10개의 노드로 이루어진 모델을 만들게 됩니다.

만약 히든레이어 없이 모델을 만들었다면 첫번째 결과, 그 숫자가 0인지 아닌지를 판단하는 수식에서 입력층 노드의 모든 노드에 가중치를 붙여서 판단하게 됩니다.

학습이 끝났을 때 우리는 다음과 같이 해석할 수 있습니다.

가중치가 높게 부여된 픽셀은 0이라고 판단하기에 중요한 픽셀이다. 

가중치가 0에 가까운 픽셀은 0이라고 판단하기에 불필요한 픽셀이다 라고요.

84개의 노드를 가진 히든 레이어를 추가하여 모델을 만들어봅시다.

그것은 중간 결과로 칼럼 84개를 가진 표가 만들어지는 것입니다.

히든레이어가 있는 상황에서는 첫번째 결과 0인지 아닌지를 판단하기 위한 수식에서 히든레이어의 모든 노드가 입력으로 사용되는 것이고 모든 노드에 가중치를 붙여 판단하게 되겠죠

이 84개의 중간 결과는 인공신경망 구조 안에서 가중치 학습을 통해 컴퓨터 스스로 만들어낸 값입니다.

우리는 앞서 칼럼은 변수이며 특징이라고 배웠습니다.

이곳에 특징 84개는 컴퓨터가 인공신경망 구조 안에서 학습을 통해 스스로 찾은 것입니다.

그래서 노드가 84개인 은닉층을 추가한 행동은 컴퓨터에게 0~9까지 중 어느 숫자인지 판단하기 위해 가장 좋은 특징 84개를 찾아달라는 뜻이 됩니다.

그래서 인공신경망은 특징 자동 추출기라는 별명이 있습니다.

지금의 코드로도 훌륭합니다. 다만 이 다음 과정을 위해 이 코드를 조금 변형해봅시다. reshape 대신 flatten 이라는 것을 사용하겠습니다.

그러면 모델의 모양은 60000, 784 가 아닌 60000, 28, 28 이 됩니다.

따라서 Input 레이어의 shape은 28, 28이 됩니다. 그리고 입력 레이어 밑에 tf.keras.layers.Flatten()(X) 라는 코드를 입력합니다.
