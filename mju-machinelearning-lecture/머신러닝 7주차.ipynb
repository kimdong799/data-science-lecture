```python
# if_else filter 를 사용하여서 보다 구체적인 함수도 표현 가능
# 반드시 if 와 else 를 함께 사용해야함.
```

```python
even_power = lambda x:x**2 if x % 2 ==0 else x
# 짝수면 x 제곱, 아니면 x이다.
print(even_power(2))
print(even_power(3))
```

```python
lower_char = lambda x:x.lower() if type(x) == str else x
# str형이라면 소문자로 아니라면 그대로 유지
# lower() 소문자로 만드는 내장함수 .lower()
print(lower_char('LOVE'))
print(lower_char(3))
```

```python
print(lower_char("MIS"))
```

```python
# map()
# 함수와 순서형 자료형(문자열, 리스트, 튜플)을 입력으로 받아서 각 원소를
# 해당 함수에 차례대로 넣어 반환되는 결과값을 묶어 돌려주는 함수
# lambda 함수와 주로 함께 사용!!
# 각 원소에 동일한 함수를 적용시킴
# map(함수이름, 순서형 자료형)
```

```python
ex = [1,2,3,4,5]
f = lambda x:x**2
print(list(map(f,ex)))
```

```python
ex = [1,2,3,4,5]
f_2 = lambda x,y: x+y
print(list(map(f_2,ex, ex)))
```

# List comprehensions

```python
# 기존 list 를 사용하여 간단히 다른 list를 만드는 기법
# 포괄적인 list, 포함되는 리트스타는 의미
# 매우 자주 사용되는 기법
# for 문과 append 함수를 사용하는 것 보다 속도가 빠름
# 대용량 자료 처리 유리
```

```python
# General function
result = []
for i in range(10):
    result.append(i)
result
```

```python
# List comprehension
result = [i for i in range(10)]
result
```

```python
# if filter 사용가능 (단 else는 사용불가)
# if filter 를 사용하는 경우, 만족하는 원소만 리스트에 포함!
```

```python
[i for i in range(10) if i % 2 ==0]
# i가 짝수일경우만 i에 0~9까지 저장
```

```python
[x**2 for x in range(10) if x % 2 != 0]
# 0~9까지의 홀수를 x값에 저장 후 제곱값 출력
```

```python
[x.upper() for x in ['a','b',1,2] if type(x) == str]
# 문자열인 경우만 대문자로 출력
```

```python
# 람다와 상당히 비슷함
```

```python
case_1 = ['A','B','C']
case_2 = ['D','E','F']
result = [i+j for i in case_1 for j in case_2]
result
# for 문이 적힌 순서대로 loop가 동작!
```

```python
# Two dimensional list(리스트 안의 리스트 구조)도 생성 가능!
case_1 = ['A','B','C']
case_2 = ['D','E','F']
result_2 = [[i+j for i in case_1] for j in case_2]
result_2
```

```python
sent = "I love you so much"
words = sent.split()
stuff = [[w.upper(), w.lower(), len(w)] for w in words]
for i in stuff :
    print(i)
```
