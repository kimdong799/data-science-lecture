```python
def check_resNum():
    resident_num = input(" !")

    pattern = re.compile("-")
    pattern2 = re.compile("[1-4]")
    pattern3 = re.compile("[0-9]")

    splited = pattern.split(resident_num)
    first_digit = splited[0]
    second_digit = splited[1]

    if len(first_digit) != 6 or len(second_digit) != 7:
        print(" .")
    elif pattern2.search(second_digit[0])==None:
        print(" .")
    else:
        subbed = re.sub("[0-9]","*",second_digit[1:])
        print(first_digit+"-"+ second_digit[0]+subbed)
```

```python
for i in range(6):
    check_resNum()
    print()
```

```python
def check_password():

    password = input(" ")

    import re
    pattern1 = re.compile("[0-9]")
    pattern2 = re.compile("[a-z]")
    pattern3 = re.compile("[A-Z]")

    if len(password) >= 8 and len(password) <15:
        if pattern1.findall(password)==[]:
            print(".")
        elif pattern2.findall(password)==[]:
            print(".")
        elif pattern3.findall(password)==[]:
            print(".")
        else:
            print(".")
    else:
        print(".")
```

```python
for i in range(5):
    check_password()
    print()
```
