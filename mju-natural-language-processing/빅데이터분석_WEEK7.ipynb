# 재추출을 통해 한 row씩 가져오자

결과는 동일하나 한 행씩 가져와서 다시 열단위로 쪼개서 출력하는 것.

```python
import requests
from bs4 import BeautifulSoup

# res = resquests.get('')
f = open("../docs/7주차.html", 'r', encoding='utf8')
soup = BeautifulSoup(f, 'html.parser')
items = soup.select("tr") # 일정, 타이틀 난이도 컬럼으로 이루어짐.

# 리스트 형태로 받아왔기 때문에 반복문 사용
for item in items:
    columns = item.select('td')
    row_str = '' # 빈 문자열 생성.
    for column in columns:
        row_str = row_str +', '+column.get_text()
    print(row_str[2:]) # 두번째부터 끝까지
```

## findall 함수: 정규표현식과 매칭되는 모든 문자열을 리스트 객체로 리턴함

# findall 함수를 사용해서 정규표현식에 해당되는 문자열이 있는지 확인하기

```python
import re
pattern = re.compile('[a-z]+')
findalled = pattern.findall("GAME")

# 리스트로 반환
if len(findalled)>0:
    print("정규표현식에 맞는 문자열이 존재")
else:
    print("정규표현식에 맞는 문자열이 존재하지 않음")
```

## split 함수: 찾은 정규표현식 패턴 문자열을 기준으로 문자열을 분리

```python
import re
pattern2 = re.compile(' [A-Z]{2} ') # 앞뒤로 스페이스가 있고 대문자가 2번 반복
splited = pattern2.split("python VS java") # 기준으로 나누어라
print(splited)
```

## sub 함수: 찾은 정규표현식 패턴 문자열을 다른 문자열로 변경

```python
import re
pattern2 = re.compile('-')
subed = pattern2.sub('*', '12345-1234567') # sub(바꿀 문자열, 본래 문자열)
print(subed)
```

```python
subed = re.sub('-','*','123456-1234567') # sub(정규표현식, 바꿀문자열, 본래문자열)
```

```python
print(subed)
```
