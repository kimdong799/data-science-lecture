```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/sise/sise_rise.nhn")
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select("a.tltle")

# 리스트 형태로 받아왔기 때문에 반복문 사용
title=[]
for item in items:
    title.append(item.get_text())
    print(item.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/sise/sise_rise.nhn")
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select("td.number")

# 리스트 형태로 받아왔기 때문에 반복문 사용
number=[]
for item in items:
    number.append(item.get_text().strip())
    print(item.get_text().strip())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(title)
df
```

```python
l=[]
n=0
m=10
for i in range(len(title)):
    l.append(number[n:m])
    n+=10
    m+=10
print(l)
```
