```python
baseUrl = "https://www.instagram.com/explore/tags/"
plusUrl = input("검색할 태그를 입력하세요 : ")
url = baseUrl + quote_plus(plusUrl)

# 인스타그램의 태그를 입력해서 URL 오픈하기

driver = webdriver.Chrome("C:/Users/Kim/Downloads/chromedriver")
driver.get(url)

# 필수요소는 아니지만 드라이버 실행하는 시간을 기다리는 것.
# 사진이 다 뜨기 전에 결과에 저장이 끝날 수 있음.
time.sleep(3) 

html = driver.page_source
soup = BeautifulSoup(html)

insta = soup.select(".v1Nh3.kIKUG._bz0w") # select를 사용 인스타의 경우 하나만 있어도 괜찮다.

#print(insta[0]) # 한개만 가져오기(속성 확인)

# a태그의 href를 통해 해당 태그 주소 가져오기
n =1
for i in insta:
    print("https://instagram.com"+i.a['href'])
    imgUrl = i.select_one(".KL4Bh").img["src"]
    with urlopen(imgUrl) as f:
        with open("../docs/img/"+plusUrl+str(n)+'.jpg', 'wb') as h:
            img = f.read()
            h.write(img)
    
    print(plusUrl+str(n)+" 다운로드 중...")
    print()
    n+=1

print("프로그램 종료")

driver.close()
```
