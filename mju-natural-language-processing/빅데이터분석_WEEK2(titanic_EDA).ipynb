```python
# 탑승자 생존 확률
```

```python
def get_probability(feature):
    Survived = df_train[df_train["Survived"]==1][feature].value_counts()
    Dead = df_train[df_train["Survived"]==0][feature].value_counts()
    total = Survived + Dead
    print("살아남은 확률: \n{} \n 죽은 확률: \n{}".format(Survived/total, Dead/total))
```

```python
features = ['Sex', 'Pclass', 'SibSp', 'Parch', 'Embarked']
for feature in features:
    get_probability(feature)
```

```python
def draw_facetgrid(feature):
    facet = sns.FacetGrid(df_train, hue="Survived", aspect=5)
    facet.map(sns.kdeplot, feature, shade=True)
    facet.set(xlim=(0, df_train[feature].max()))

    facet.add_legend() 
    plt.show()
```

```python
draw_facetgrid("Age")
```
