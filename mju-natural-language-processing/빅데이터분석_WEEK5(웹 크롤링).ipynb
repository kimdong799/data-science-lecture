```python
res = requests.get("http://corners.gmarket.co.kr/Bestsellers")
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
product_name = []
mydata = soup.find_all('a', "itemname")
for i in mydata:
    print(i.get_text())
    product_name.append(i.get_text())
```

```python
product_price=[]
mydata = soup.find_all('div', "s-price")
for i in mydata:
    print(i.get_text())
    product_price.append(i.get_text())
```

```python
product_name
```

```python
product_price
```

```python
product_dict = [i for i in zip(product_name, product_price)]
product_dict
```

```python
import pandas as pd
```

```python
product_df = pd.DataFrame(product_dict)
```

```python
product_df
```

```python
product_df.head(10)
```

```python
product_df = product_df.drop([0,1,2,3,4])
```

```python
product_df = product_df.reset_index()
```

```python
product_df.drop('index', axis=1, inplace = True)
```

```python
product_df
```

```python
product_df.columns = ['','']
```

```python
product_df
```

```python
product_df[""] = product_df[""].str.split("").str[0]
```

```python
product_df[""] = product_df[""].str.replace("[^0-9 ]","") 
```

```python
product_df
```

```python
product_df
```

```python
product_df[''] = product_df[''].str.split("", expand=True)
```

```python
product_df[''] = product_df[''].str.replace(',', '').astype('int64')
```
