# shape, ndim, size

```python
arr = np.array([1, 4, 5, 8]) # 1차원 배열 (벡터)

mat = np.array([[1, 2, 5, 8], # 다차원배열 (매트릭스)
              [1, 2, 5, 8],
              [1, 2, 5, 8]])
```

```python
print("vector shape : ",arr.shape) # array의 shape을 반환

print("vector ndim : ",arr.ndim) # array의 차원을 반환

print("vector size : ",arr.size) # array의 elements 개수를 반환
```

```python
print("matrix shape : ",mat.shape) # array의 shape을 반환

print("matrix ndim : ",mat.ndim) # array의 차원을 반환

print("matrix size : ",mat.size) # array의 elements 개수를 반환
```

# Handling shape, reshape

```python
# 2행 4열의 2차원 배열 생성

test_mat = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(test_mat.shape)
print(test_mat.ndim)
```

### reshape

```python
reshape_mat = test_mat.reshape(8, ) #  (8, ) -> 1차원 배열로 변환해라
print(reshape_mat)
print(reshape_mat.shape)
```

```python
reshape_mat.reshape(2, 4) # 다시 2행 4열로 바꿔보자
```

```python
test_mat.reshape(-1, 2) # -1은 자동으로 값을 할당 -1, 2-> n행 2열로 반환 = 원소의 갯수에 따라 자동생성
```

### flatten

```python
test_mat.flatten() # flatten 함수는 다차원 array를 1차원으로 변환시켜준다.
```

# all & any

## all (and), any (or) 개념

```python
a = np.arange(10)
a
```

```python
np.any(a>5), np.any(a<0) # 하나라도 조건을 만족하면 True 반환
```

```python
np.all(a>5), np.all(a<10) # 모든 원소가 조건을 만족하면 True 반환
```

```python
a = np.array([1, 3, 0])
b = np.array([5, 2, 1])
```

```python
a>b # 각각의 원소별로 비교,  행렬간의 비교는 각각의 원소단위에서 이루어진다.
```

```python
a==b # 다른 연산자를 활용해도 마찬가지이다!
```

```python
# 비교연산자를 통해 array간의 비교가 가능하고 반환값 역시 array이기 때문에 반환값에 all, any 함수를 사용 가능하다.
```

```python
(a>b).any() # a>b 의 결과값 중 하나라도 true라면 true 반환
```

```python
(a>b).all() # a>b 의 결과값 모두가 true라면 true 반환
```

# where

where 함수의 경우 원핫인코딩이나 더미변수 생성에 용이하게 사용 가능하다.

```python
a = np.arange(1, 11)
a
```

# boolean index

```python
# True, False 배열을 반환

test_array = np.array([1, 4, 0, 2, 3, 8, 9, 7], float)
test_array > 3
```

```python
# test_array 에서 3을 초과하는 원소들의 배열을 반환

test_array[test_array>3]
```

```python
# 또 다른 방법

condition = test_array < 3

test_array[condition]
```

```python
a = np.arange(10)
a[a>5] = 0

print("60161051 김동현\n")
print(a)
```

```python
b = np.arange(1, 20, 2) 
condition_index = np.where((b>4)&(b<14)) # where 함수를 통해 조건에 해당하는 index array생성

for i in condition_index: # 해당 index 접근, 값 변경
    b[i] = b[i] * -1
print("60161051 김동현\n")
print(b)
```

```python
np.flip(b) # flip 함수 사용
```

```python
c = np.ones(12, int).reshape(3, 4) # ones 함수사용, 1로 이루어진 3*4 배열 생성(reshape)
print(c)
print("----------------")
c = c.reshape(2, 6) # 2행 6열로 변경
print(c)
print("----------------")
c = c.transpose() # 행열 전치 transpose 함수
print(c)
print("----------------")
c.sum() # 모든 원소의 합
```
