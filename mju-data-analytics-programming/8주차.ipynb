```python
import csv

with open('../data/tmp_csv.csv', 'w', encoding='utf-8-sig', newline='') as writer_csv:
    field_name_list = ['First Name', 'Last Name'] #  
    writer = csv.DictWriter(writer_csv, fieldnames=field_name_list) #     
    writer.writeheader() #  csv        writerheader()    
    writer.writerow({'First Name': 'Dave', 'Last Name':'Lee'})
    writer.writerow({'First Name': 'Dave', 'Last Name':'Kim'})
    writer.writerow({'First Name': 'Robin', 'Last Name':'Park'})
```

## branch_name, address   ?

```python
data_file = open('../data/parkinglot.xml', 'r', encoding='utf-8-sig')
soup = BeautifulSoup(data_file, 'xml')

items = soup.select('item')

for item in items:
    branch_name = item.select_one('branch_name')
    print(': ',branch_name.text)
    address = item.select_one('address')
    print(': ',address.text)
```

```python
branch_name = []
address = []

data_file = open('../data/parkinglot.xml', 'r', encoding='utf-8-sig')
soup = BeautifulSoup(data_file, 'xml')

items = soup.select('item')

for item in items:
    branch_name.append(item.select_one('branch_name'))
    address.append(item.select_one('address'))

data_file.close()
```

```python
#      csv    

import csv

with open('../data/week8_hw.csv', 'w', encoding='utf-8-sig', newline='') as writer_csv:
    field_name_list = ['', ''] #  
    writer = csv.DictWriter(writer_csv, fieldnames=field_name_list) #     
    writer.writeheader() #  csv        writerheader()    
    
    for i, j in zip(branch_name, address):
        writer.writerow({'': i.text, '':j.text})
```
