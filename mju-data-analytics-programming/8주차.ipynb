## 예시 (한줄만 출력해보자)

```python
import csv

data_file = open('../data/train.csv', 'r', encoding='utf-8-sig')
data_lines = csv.reader(data_file, delimiter=',')

# 0은 첫번째 컬럼을 의미
for line in data_lines:
    print(line) # 첫번째 row
    print()
    print(line[0]) # 첫번째 컬럼
    break # 편의상 한줄만
    
data_file.close() # 파일은 항상 닫아주자
```

## csv 파일 쓰기
- open 시 'w'로 옵션을 설정
- open() 함수에 newline=''을 넣어주는 이유는 윈도우의 경우에만 csv 모듈에서 데이터를 쓸 때 각 라인 뒤에 빈 라인이 추가되는 문제가 있기 때문에 이를 없애기 위해 파일을 open 할 때 newline=''와 같은 옵션을 지정

- csv.reader 대신 csv.writer 함수 사용

- writerow 함수에 리스트 형식으로 데이터를 넣으면 한 라인씩 파일에 추가됨
- 파일 close는 파일 읽기와 동일함

```python
import csv

data_file = open('../data/tmp_csv.csv', 'w', encoding='utf-8-sig', newline='')
data_write = csv.writer(data_file, delimiter=',')
data_write.writerow(['1', '2', '3'])
data_file.close()
```

## CSV 파일 쓰기 다른 기법(with)
- with 문법을 사용해서 with 내부 구문 실행 완료 후 자동으로 파일을 닫을 수 있음

```python
import csv
with open('../data/tmp_csv.csv', 'w', encoding='utf-8-sig', newline='') as writer_csv:
    writer = csv.writer(writer_csv, delimiter=',')
    writer.writerow(['love']*3 + ['banana'])
    writer.writerow(['apple', 2])
    writer.writerow(['Spam', 'Lovely Spam', 'Wonderful Spam'])
```

## csv 파일 쓰기 다른 기법(딕셔너리 타입으로 파일 쓰기)
- csv.writer 함수 대신 csv.DictWriter 함수 사용
- field 이름 선언 후 데이터넣기

```python
import csv

with open('../data/tmp_csv.csv', 'w', encoding='utf-8-sig', newline='') as writer_csv:
    field_name_list = ['First Name', 'Last Name'] # 필드명 정의
    writer = csv.DictWriter(writer_csv, fieldnames=field_name_list) # 필드명을 미리 선언할 수 있음
    writer.writeheader() # 보통 csv 파일 상단에는 필드명을 넣기 때문에 선언된 필드명을 writerheader() 함수로 넣을 수 있음
    writer.writerow({'First Name': 'Dave', 'Last Name':'Lee'})
    writer.writerow({'First Name': 'Dave', 'Last Name':'Kim'})
    writer.writerow({'First Name': 'Robin', 'Last Name':'Park'})
```

```python
# 앞서 저장한 csv 파일 읽어보기

with open('../data/tmp_csv.csv', 'r', encoding='utf-8-sig') as reader_csv:
    reader = csv.reader(reader_csv, delimiter=',')
    
    for row in reader:
        print(row)
```

## XML 포맷 이해하기

- XML(Extensible Markup Language)
    - 특정 목적에 따라 데이터를 태그로 감싸서 마크업하는 범용적인 포맷
    - 마크업 언어는 태그 등을 이용하여 데이터의 구조를 기술하는 언어의 한 가지
    - 가장 친숙한 마크업 언어가 HTML
    - XML은 HTML과 마찬가지로 데이터를 계층 구조로 표현
    - XML 기본 구조
    
        <태그 속성 = "속성값">내용</태그>
        
        
    - 태그와 속성은 특정 목적에 따라 임의로 이름을 정해서 사용
    
        <product id="M001" price="300000">32인치 LCD 모니터</product>
        
    - 다른 요소와 그룹으로 묶을 수도 있음
    
        <products type="전자제품">
            <product id="M001" price="300000">32인치 LCD 모니터</product>
            <product id="M002" price="210000">24인치 LCD 모니터</product>
        </products>

XML의 계층화된 구조는 CSV나 plain text로는 나타내기 제한된다. 이것이 XML의 장점

## branch_name, address로 구성된 목록을 만든다면?

```python
data_file = open('../data/parkinglot.xml', 'r', encoding='utf-8-sig')
soup = BeautifulSoup(data_file, 'xml')

items = soup.select('item')

for item in items:
    branch_name = item.select_one('branch_name')
    print('지점명: ',branch_name.text)
    address = item.select_one('address')
    print('주소: ',address.text)
```

```python
branch_name = []
address = []

data_file = open('../data/parkinglot.xml', 'r', encoding='utf-8-sig')
soup = BeautifulSoup(data_file, 'xml')

items = soup.select('item')

for item in items:
    branch_name.append(item.select_one('branch_name'))
    address.append(item.select_one('address'))

data_file.close()
```

```python
# 앞선 예제에서 진행한 주차 데이터를 csv 파일 형태의 데이터셋으로 변환

import csv

with open('../data/week8_hw.csv', 'w', encoding='utf-8-sig', newline='') as writer_csv:
    field_name_list = ['지점명', '주소'] # 필드명 정의
    writer = csv.DictWriter(writer_csv, fieldnames=field_name_list) # 필드명을 미리 선언할 수 있음
    writer.writeheader() # 보통 csv 파일 상단에는 필드명을 넣기 때문에 선언된 필드명을 writerheader() 함수로 넣을 수 있음
    
    for i, j in zip(branch_name, address):
        writer.writerow({'지점명': i.text, '주소':j.text})
```

```python
# 공공데이터포털 담배소매형환 데이터

from bs4 import BeautifulSoup

data_file = open('../data/담배소매현황.xml', 'r', encoding='utf-8-sig')
soup = BeautifulSoup(data_file, 'xml')

items = soup.select('CONT_DATA_ROW')

for item in items:
    SHD_NM = item.select_one('SHD_NM')
    print('지점명: ',SHD_NM.text)
    CO_F2 = item.select_one('CO_F2')
    print('주소: ',CO_F2.text)
```
