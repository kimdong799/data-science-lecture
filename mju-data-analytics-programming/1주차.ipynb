```python
from turtle import *

color('red', 'yellow')
begin_fill()
while True:
    forward(200)
    left(170)
    if abs(pos()) < 1:
        break
end_fill()
done()
```

```python
# 반복문으로 삼각형을 그려보자!

t.clear()


t.begin_fill() # 도형의 색을 채우자
for i in range(3):
    t.forward(100)
    t.left(120)
t.end_fill()
```

```python
# 반복문으로 삼각형을 그려보자!

t.clear()
t.home()


t.begin_fill() # 도형의 색을 채우자
for i in range(3):
    t.forward(100)
    t.left(110) # 각도가 약간 맞지 않아도 색이 칠해짐
t.end_fill()
```

```python
# 경영정보학과 60161051 김동현 1주차 과제(데프)

n = int(input("숫자를 입력하세요: "))
color = input("색상을 입력하세요: ")

t.clear()
t.home()
t.color(color) # 색상 지정

t.begin_fill() # 도형의 색을 채우자
for i in range(n):
    t.forward(100)
    t.left(360/n) # 각도가 약간 맞지 않아도 색이 칠해짐
t.end_fill()
```
