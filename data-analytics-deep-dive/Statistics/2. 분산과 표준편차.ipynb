## 기댓값과 분산의 설명

기댓값은 확률변수에서 어떤 값이 나올지 예측한 것이라면 분산은 그 예측값(기댓값)이 어느정도의 오차를 가질지 예측의 정확성, 신뢰성을 정량적으로 나타낸 값이라 할 수 있다.

## 확률분포의 분산

확률분포의 분산 공식은 다음과 같다.

$\sigma^2 = Var[X] = E[(X - \mu)^2]$

이산확률변수의 분산은 평균($\mu$)으로부터 표본데이터까지의 거리의 제곱을 확률질량함수로 가중합한 값이다.

\begin{align}
\sigma^2 = \sum_{x_i \in \Omega} (x_i - \mu)^2 p(x_i)
\end{align}

연속확률변수의 분산은 평균($\mu$)로부터 표본데이터까지의 거리의 제곱을 확률밀도함수로 가중하여 적분한 값이다.

\begin{align}
\sigma^2 = \int_{-\infty}^{\infty} (x - \mu)^2 p(x)dx
\end{align}

## 두 확률변수의 합의 분산

두 확률변수 $X, Y$의 합의 분산은 각 확률변수의 분산의 합과 다음과 같은 관계가 성립한다.

$Var[X+Y] = Var[X] + Var[Y] + 2E[(X - \mu_x)(Y-\mu_y)]$

## 확률변수의 독립

두 확률변수가 독립이라는 말은 쉽게 두 확률변수가 서로에게 영향을 미치지 않는다고 할 수 있다.

반대로 종속이라는 말은 두 확률변수가 서로의 확률분포에 영향을 미치는 것이다.

종속인 경우

```python
x = [60, 70, 80, 90, 100]
y = [150, 160, 170, 180, 190]

plt.plot(x, y, 'ro')
plt.xlim(40, 120)
plt.ylim(130, 210)
plt.xlabel('몸무게')
plt.ylabel('키')
```

독립인 경우

```python
x = [100, 100, 100, 100, 100]
y = [150, 160, 170, 180, 190]

plt.plot(x, y, 'ro')
plt.xlabel('IQ')
plt.ylabel('키')
```

이처럼 두 확률변수 $X, Y$가 독립이면 다음 식이 성립한다.

$Var[X+Y] = Var[X] + Var[Y] + 2E[(X - \mu_x)(Y-\mu_y)]$

위 식에서 마지막 항

이 $E[(X - \mu_x)(Y-\mu_y)] = 0$가 된다.

이 등식을 따라서 서로 독립인 경우의 두 확률변수의 합의 분산은 각 확률변수의 분산의 합이된다.

```python
x = [100, 100, 100, 100, 100]
y = [150, 160, 170, 180, 190]

np.var(x)+np.var(y)
```

## 표본평균의 분산

이전에 확률변수 $X$의 기댓값과 표본평균 $\bar{X}$의 기댓값은 같다는 것을 증명했다.

$E[\bar{X}] = E[X]$

표본평균 $\bar{X}$의 분산 $Var[\bar{X}]$는 원래 확률변수 $X$의 분산 $Var[X]$와 다음의 관계가 성립한다.

$Var[\bar{X}] = \dfrac{1}{N}Var[X]$

즉, 원래 확률변수 $X$의 분산에서 더 작아진 값이 표본평균의 분산이다.

아래의 시각화 결과를 확인해보자.

```python
x = np.random.randn(1000)

x_bar = []
for i in range(100):
    x_bar.append(np.mean(np.random.randn(1000)))

plt.figure(figsize=(8, 5))
plt.subplot(121)
sns.distplot(x)
plt.xlim(-5, 5)
plt.subplot(122)
sns.distplot(x_bar, color='orange')
plt.xlim(-5, 5)
plt.show()
```

```python
np.var(x), np.var(x_bar)
```

실제로 표본평균의 분산은 기존 확률변수의 분산보다 확연히 좁아진 모습을 보인다.

> 따라서 표본평균을 계산한 표본의 갯수가 커지면 표본평균의 변동은 작아진다.

즉, 데이터가 무한히 많다면 결국 $\bar{X}$는 원래 확률변수 $X$의 기댓값이 된다.

## 표본분산의 기댓값

표본평균의 기댓값은 같다는 것을 위에서 증명했다.

하지만 표본분산의 기댓값은 표본분산과 같지 않고 $\dfrac{N-1}{N}\sigma^2$이 된다.

주피터에서의 증명은 생략

최종적으로 표본분산의 기댓값이 정확하게 $\sigma^2$이 되려면 평균과의 거리의 제곱에 $\dfrac{1}{N}$이 아닌 $\dfrac{1}{N-1}$을 곱해야한다.

이것을 비편향 표본분산이라 한다.

\begin{align}
S^2_{\text{unbiased}} \equiv \dfrac{1}{N-1} \sum (X_i-\bar{X})^2
\end{align}

즉, 표본분산보다 더 작은 값이 비편향표본분산(실제분산)이 된다.

그 이유는 다음과 같다.

1. 표본분산을 계산할 때 사용되는 표본평균값은 데이터가 몰려있는 쪽으로 더 편향된다.

2. 따라서 편향된 표본평균을 이용하기에 각 데이터까지의 거리를 계산하면 원래의 기댓값으로부터의 거리보다 더 작게 나올 수 있다.

```python
 np.random.seed(15)
N = 7

data = np.sort(np.random.normal(size=(N)))[::-1]
data
```

```python
plt.plot(data, np.zeros(7), 'bo')
plt.grid()
```

```python
mean = np.mean(data)
mean
```

이 경우 음수인 데이터가 더 많기 때문에 표본평균값이 음수가 된 상황이다.

```python
sns.rugplot(data, height=0.5, linewidth=4)
x = np.linspace(-3, 3, 100)
plt.axvline(x=0, ls=':', c='r', linewidth=2, label='실제 기댓값')
plt.axvline(x=mean, ls='--', c='b', linewidth=2, label='표본평균')
plt.legend()
plt.xlim(-2, 2)
plt.grid()
plt.show()
```

데이터의 위치와 표본평균 위치를 시각화 한 결과이다.

각 샘플데이터와 평균과의 거리이다.

```python
distance = data - mean
distance
```

각 거리의 제곱에 평균을 구하면 표본분산의 기댓값이 된다.

```python
sample_variance = (distance ** 2).mean()
sample_variance
```

비편향 표본분산으로 구한 값보다 더 작은 값이 되는 것을 확인할 수 있다.

```python
sample_variance * N / (N-1)
```

**주의할 점은 표본분산의 기댓값이 원래의 분산보다 작은 값이 나오는 경향이 있다는 것이다. 항상 원래의 분산값보다 작게 나오는 것은 아니다.**
