## 적분

적분(integral)은 미분과 반대되는 개념으로 널리 알려져있다. 하지만 실제 적분에는 두가지가 있다.

1. 부정적분(indefinite integral)
2. 정적분(definite integral)



## 부정적분

부정적분(indifinite integral)은 앞의 적분에 대한 설명과 같이 정확하게 미분과 반대되는 개념이다. 즉, 반-미분(anti-derivative)라고 할 수 있다.

어떤 함수$f(x)$를 미분한 결과인 도함수를 다시 미분되기 전의 함수로 복원하는 과정 또는 그 결과를 부정적분이라한다.

\begin{align}
\dfrac{dF(x)}{dx} = f(x) \;\;\leftrightarrow\;\; F(x) = \int_{}^{} f(x) dx + C 
\end{align}

위 식에서 $C$는 상수항을 뜻한다. 왜냐하면 도함수는 상수항을 제거하기 때문에 원래 함수에 어떤 상수항이 존재할 수 있기 때문이다. 따라서 무한 개의 해가 존재하므로 상수항 $C$를 추가하며 매우 당연하기에 생략하는 경우도 존재한다.

## 편미분의 부정적분

편미분을 한 도함수에서 원래의 함수를 찾을 수 있다. 

$f(x, y)$가 원래의 함수를 어떻게 미분했는지에 따라 원래의 함수를 표기하는 방법이 달라진다.

만약 $x$로 편미분한 경우 다음과 같다.

\begin{align}
\dfrac{\partial F_1(x, y)}{\partial x} = f(x, y) \; \leftrightarrow \; F_1(x, y) = \int_{}^{} f(x, y) dx + C(y)
\end{align}

여기서 중요한 점은 마지막 항 $C(y)$는 $y$와 관련한 함수 혹은 일반 상수일 수 있다.

이는 $x$로 편미분 된 도함수에 대해 적분을 하는 것이기에 알 수 없다.

## 다차 도함수와 다중적분

미분을 여러 번 한 결과로 나온 다차 도함수로부터 원래의 함수를 찾아내기 위해선 적분도 마찬가지로 여러 번 수행해야 한다. 이를 다중적분이라한다.

예를 들어 $f(x, y)$가 함수 $F_3(x, y)$를 $x$로 한번, $y$로 한번 편미분한 이차 도함수라 가정하자.

이 이차도함수의 원래 함수를 찾기 위해선 $y$로 한번, $x$로 한번 적분 해야한다. 

\begin{align}
\dfrac{\partial^2 F_3(x,y)}{\partial x \partial y} 
= f(x, y) \; \leftrightarrow \; F_3(x, y) 
= \int_x \int_y f(x, y) dydx
\end{align}

적분 기호 아래의 변수명을 생략하고 다음처럼 표기할 수 있다.

\begin{align}
\iint f(x, y) dydx
\end{align}

## 정적분

정적분은 독립변수 $x$가 어떤 구간 $[a, b]$사이일 때 그 구간에서 함수 $f(x)의 값과 수평선 $x$축이 이루는 면적 구하는 행위 혹은 그 값을 말한다.

\begin{align}
\int_{a}^{b} f(x) dx 
\end{align}

```python
from matplotlib.patches import Polygon


def f(x):
    return x ** 3 - 3 * x ** 2 + x + 6


a, b = 0, 2
x = np.linspace(a - 0.5, b + 0.5, 50)
y = f(x)

ax = plt.subplot(111)
plt.title("정적분의 예")
plt.plot(x, y, 'r', linewidth=2)
plt.ylim(bottom=0)
ix = np.linspace(a, b)
iy = f(ix)
verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 0.2 * (f(a) + f(b)), r"$\int_a^b f(x)dx$",
         horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.05, '$x$')
plt.figtext(0.1, 0.9, '$y$')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([])
ax.set_xlim(-2, 4)
ax.set_ylim(0, 8)
plt.show()
```

위 그래프의 면적을 구하는 행위는 부정적분으로 구한 함수 $F(x)$를 이용하면 다음처럼 정적분의 값을 구할 수 있다.

\begin{align}
\int_{a}^{b} f(x) dx = F(b) - F(a)  
\end{align}

이를 미적분학의 기본 정리 라고 부른다.

정적분은 심파이 등으로 부정적분을 한 뒤 미적분학의 기본정리를 이용하여 푸는 방법

원래 함수의 면적 부분을 실제로 잘게 쪼개어 면적을 근사하게 구하는 수치적분

두가지 방법으로 구할 수 있다.

다음 정적분을 구한다고 가정해보자.

\begin{align}
\int_0^2 ( x^3 - 3x^2 + x + 6) dx
\end{align}

```python
x, y = sympy.symbols('x y')
f = x ** 3 - 3 * x ** 2 + x + 6
f
```

위 함수의 부정적분을 구하면 다음과 같다.

```python
# 부정적분

F = sympy.integrate(f)
F
```

이렇게 구해진 미분하기 전의 함수에 정적분 구간을 넣어주면 된다.

subs(), evalf() 메소드를 이용한다.

```python
(F.subs(x, 2) - F.subs(x, 0)).evalf()
```

## 수치적분

수치적분은 앞의 소개에서 함수를 아주 작은 구간으로 나누어 실제 면적을 계산하여 정적분을 구하는 방법이다. 

사이파이의 quad() 명령으로 가능하다.

```python
def f(x):
    return x ** 3 - 3 * x ** 2 + x + 6

import scipy as sp

sp.integrate.quad(f, 0, 2)
```

결괏값의 두번째는 오차의 상한값을 뜻한다.

아래의 두 함수를 부정적분, 수치적분 두가지 방법으로 구해보자.

\begin{align}
\int_0^1 (3x^2 - 6x + 1)d
\end{align}

```python
# 부정적분으로 정적분을 구하는 방법

x = sympy.symbols('x')
f = 3 * x ** 2 - 6 * x + 1
f
```

```python
F = sympy.integrate(f, x)
F
```

```python
(F.subs(x, 1) - F.subs(x, 0)).evalf()
```

```python
# 수치적분으로 구하는 방법

def f(x):
    return 3 * x ** 2 - 6 * x + 1

sp.integrate.quad(f, 0, 1)
```

\begin{align}
\int_1^{10} \left( 2 + 6x + 4\exp(x) + \dfrac{5}{x} \right) dx
\end{align}

```python
# 부정적분으로 구하는 방법

x = sympy.symbols('x')
f = 2 + 6 * x + 4 * sympy.exp(x) + 5/x
f
```

```python
F = sympy.integrate(f, x)
F
```

```python
(F.subs(x, 10)-F.subs(x, 1)).evalf()
```

```python
# 수치적분으로 구하는 방법

def f(x):
    return 2 + 6 * x + 4 * sympy.exp(x) + 5/x

sp.integrate.quad(f, 1, 10)
```

## 다변수 정적분

입력 변수가 2개인 2차원 함수 $f(x, y)$의 경우 정적분을 다양한 방법으로 정의할 수 있다.

예를 들어 다음과 같이 두 변수로 모두 적분하는 것은 2차원 평면에서의 주어진 사각형 영역 아래의 부피를 구하는 것과 같다.

\begin{align}
\int_{y=c}^{y=d} \int_{x=a}^{x=b} f(x, y) dx dy 
\end{align}

다음 함수는 정적분을 이용해 함수의 부피를 구하는 모습을 시각화 한 결과이다.

\begin{align}
\int_{y=0}^{y=6}\int_{x=2}^{x=8} x^2 - 10 x + y + 50 \;\; dx dy
\end{align}

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
_x = np.arange(12) / 2 + 2
_y = np.arange(12) / 2
X, Y = np.meshgrid(_x, _y)
x, y = X.ravel(), Y.ravel()
z = x * x - 10 * x + y + 50
z0 = np.zeros_like(z)
ax.bar3d(x, y, z0, 0.48, 0.48, z)
ax.set_xlim(0, 10)
ax.set_ylim(-2, 10)
ax.set_zlim(0, 50)
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.title("f(x, y)")
plt.show()
```

수치이중적분을 위해서 사이파이의 integrate 서브패키지의 dblquad() 명령을 이용한다.

`dblquad(func, a, b, gfunc, hfunc)`

a, b 는 x의 하한, 상한을 의미하고 gfun, hfun은 y의 상한 하한을 나타낸다.

여기서 gfun, hfun은 x의 함수여야 한다.

예를 들어 다음의 정적분을 수치적분으로 계산하기 위해 dblquad() 함수를 사용할 수 있다.

\begin{align}
\int_0^{\infty} \int_1^{\infty} \dfrac{\exp(-xy)}{x^2} dx dy
\end{align}

```python
def f(y, x):
    return np.exp(-x * y) / x**2

sp.integrate.dblquad(f, 1, np.inf, lambda x:0, lambda x:np.inf)
```

### 예제

다음의 정적분을 수치적분으로 구해보자.

\begin{align}
\int_{-1}^1 \int_{-1}^1 \left( 1 + xy \right) dxdy
\end{align}

```python
def f(x, y):
    return 1 + x*y

sp.integrate.dblquad(f, -1, 1, lambda x:-1, lambda x: 1)
```

## 다차원 함수의 단일 정적분

2차원 함수이지만 이중적분이 아닌 단일 정적분을 수행하는 경우도 존재한다.

이 때는 하나의 변수만 진짜 변수로 보고 나머지 하나는 상수라고 간주한다.

\begin{align}
\int_a^b f(x, y) dx
\end{align}

\begin{align}
f(x, y) = 4x^2 + 4xy + y^2
\end{align}

위 함수에서 변수 $x$만 진짜 변수로 본다면 $y$는 상수 취급되며 다음과 같은 1차원 함수가 된다.

\begin{align}
f(x; y) = 4x^2 + (4y)x + (y^2)
\end{align}
