## 선형종속과 선형독립

선형종속과 선형독립을 설명하기 전에 선형조합에 대해 다시 한 번 살펴보자.

선형조합이란 벡터 $v_1, v_2, v_3 ... v_n$에 대해 스칼라 $c_1, c_2, c_3, ..., c_n$ 곱을 모두 더한 것이다.

선형조합(Linear combination) = $v_1c_1 + v_2c_2 + ... + v_nc_n$

이러한 선형조합에서 선형조합의 결과가 영벡터 $0_n$가 되도록 하는 스칼라 계수가 존재하면 이 벡터를 **선형종속(Linearly dependent)**이라 한다.

단, 스칼라 계수가 모두 0인 경우는 제외한다.

반대로 **선형독립(Linearly independent)**이란 선형조합의 결과가 영벡터가 되는 계수들이 영벡터 이외에는 존재하지 않을 때, 선형적으로 독립이라고 한다.

\begin{align}
c_1 x_1 + \cdots + c_N x_N = 0  \;\; \rightarrow \;\; c_1 = \cdots = c_N = 0  
\end{align}

$$or$$

\begin{align}
c_1 x_1 + \cdots + c_N x_N = 0  \;\; \leftrightarrow \;\; c_1 = \cdots = c_N = 0  
\end{align}

예를 들어 임의의 선형독립인 벡터 $x_1과 x_2$가 있다고 가정하자.

\begin{split} 
\begin{align}
x_1 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \;\;
x_2 = \begin{bmatrix} 3 \\ 3 \end{bmatrix}
\end{align}
\end{split}

위 벡터가 선형독립이기 위해선 두 벡터에 대한 선형조합의 결과가 0이 되는 스칼라 계수가 0벡터 이외에는 존재하지 않아야한다.

위의 두 벡터는 1:2, 3:3 으로 그 비율이 다르다는 것을 한 눈에 파악할 수 있다.
실제 좌표 평면도위에 그려봐도 두 벡터의 방향이 다르다는 것을 알 수 있다.

따라서 두 벡터는 방향이 다르므로 아무리 두 벡터의 길이를 조절해도(스칼라 곱을 해도) 두 벡터의 조합은 0이 될 수 없다.

그러므로 위의 두 벡터는 선형독립이라 할 수 있다.

이번에는 선형종속인 경우를 살펴보자.

\begin{split} 
\begin{align}
x_1 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \;\;
x_2 = \begin{bmatrix} 3 \\ 3 \end{bmatrix}, \;\;
x_3 = \begin{bmatrix} 5 \\ 7 \end{bmatrix}
\end{align}
\end{split}

위의 벡터들을 잘 조합하면 아래의 식으로 선형조합의 결과를 0으로 반들수 있다.

$$2x_1 + x_2 - x_3 = 0$$

numpy로 계산해도 그 결과는 선형종속이다.

```python
x1 = np.array([1,2])
x2 = np.array([3,3])
x3 = np.array([[5,7]])
```

```python
x1 * 2 + x2 == x3
```

## 선형독립과 선형 연립방정식

선형독립 관계를 행렬과 벡터의 곱으로 다음과 같이 나타낼 수 있다.

\begin{split} 
\begin{align}
c_1 x_1 + \cdots + c_N x_N =
\begin{bmatrix} x_1 & x_2 & \cdots & x_N \end{bmatrix} 
\begin{bmatrix} c_1 \\ c_2 \\ \vdots \\ c_N \end{bmatrix}
= X c
\end{align}
\end{split}

위의 식을 통해 $Xc = 0$이 성립하면 선형독립이므로 선형독립인지 아닌지를 알아내는 문제는 선형 연립방정식을 푸는 문제와 같다는 것을 알 수 있다.

## 선형종속인 경우

1. 벡터의 개수가 벡터의 차원보다 크면 선형종속이다.

따라서 영벡터가 아닌 해도 존재한다는 뜻이 되며 위와 같은 벡터들은 선형종속이라 할 수 있다.

이 경우 중복된 데이터가 있으므로 선형종속이라 할 수 있다.

3. 어떤 벡터가 다른 벡터의 선형조합이면 반드시 선형종속이다.

위의 식이 성립하므로 선형종속이다.

## 랭크

행렬의 열벡터 중 서로 독립하는 열벡터의 최대 개수를 **열랭크(column rank)**라고 한다.

헹벡터 중 서로 독립인 행벡터의 최대 개수를 **행랭크(row rank)**라고 한다.

행랭크와 열랭크는 아래의 정의가 성립한다는 점을 기억하자.

**행랭크와 열랭크는 항상 같다.**

위의 정의에 따라 통상적으로 그냥 랭크라고도 한다.

따라서 행의 개수 N, 열의 개수 M인 행렬에서의 랭크는 N과 M 중 작은 값보다 커질 수 없다.

예를 들어 5 * 7 shape을 가진 행렬의 경우 랭크는 5보다 큰 값을 가질 수 없다.

\begin{align}
\text{rank} A \leq \min(M, N)
\end{align}

예를 들어 아래의 행렬에 대해 RANK를 구해보자.

\begin{split} 
\begin{align}
X_2 = \begin{bmatrix} 1 & 3 & 5 \\ 2 & 3 & 7 \end{bmatrix} 
\end{align}
\end{split}

위 행렬의 행랭크(row rank)는 간단하게 행벡터로 나누어 서로 다른 방향을 가리키므로 2인것을 알 수 있다.

반대로 열벡터로 나누어보면 [1, 2], [3, 3]은 다른 방향을 가리키지만 2 * [1, 2] + [3, 3] = [5, 7]이 되므로 열랭크는 2인것을 알 수 있다.

```python
X2 = np.array([[1, 3, 5],[2, 3, 7]])
X2
```

```python
np.linalg.matrix_rank(X2)
```

## 풀랭크

위의 예시에서 X2 행렬의 rank는 2로 행의 갯수와 같음을 확인했다.

이처럼 랭크값이 행렬의 행의 개수나 열의 개수 중 작은 값과 같으면 **풀랭크(full rank)**라 한다.

따라서 선형독립인 벡터들을 이용해 행 또는 열로 가지는 행렬을 만들면 정의에 따라 항상 풀랭크가 된다는 것을 알 수 있다.

## 로우 랭크 행렬

N차원 벡터 $x$ 하나를 이용해 만들어지는 행렬을 **랭크-1 행렬(rank-1 matrix)**라고 한다.

\begin{align}
xx^T \in \mathbf{R}^{N \times N} 
\end{align}

```python
x = np.array([[1],[0],[1]])
x
```

\begin{split} 
\begin{align}
\begin{aligned}
xx^T 
&= x \begin{bmatrix} x_1 & x_2 & \cdots & x_n \end{bmatrix} \\
&= \begin{bmatrix} x_1x & x_2x & \cdots & x_nx \end{bmatrix} 
\end{aligned}
\end{align}
\end{split}

```python
x @ x.T
```

이러한 행렬의 특징은 열벡터들이 $x$라고 하는 하나의 벡터를 $x_1배, x_2배,...x_n배$한 벡터이므로 독립적인 열벡터는 1개이다. 따라서 이를 **랭크-1 행렬**이라 부르며 랭크는 1이다.

```python
np.linalg.matrix_rank(x @ x.T)
```

이번에는 선형독립인 두 개의 N차원 벡터 $x_1, x_2$를 이용해 행렬을 만들어보자.

```python
x1 = np.array([[1],[2]])
x2 = np.array([[3],[3]])

mat = x1 @ x1.T + x2 @ x2.T
mat
```

\begin{split}
\begin{align}
\begin{bmatrix} x_1 & x_2 \end{bmatrix}
\begin{bmatrix} x_1^T \\ x_2^T \end{bmatrix}
= x_1x_1^T + x_2x_2^T
\end{align}
\end{split}

```python
np.linalg.matrix_rank(mat)
```

랭크-1 행렬과 비슷한 방법으로 행렬이 만들어지며 랭크-2 행렬의 랭크값은 2인 것을 확인할 수 있다.

이처럼 M개의 N차원 벡터를 이용하면 **랭크-M 행렬(rank-M matrix)**가 된다.

\begin{split}
\begin{align}
\begin{bmatrix} x_1 & x_2 & \cdots & x_M \end{bmatrix}
\begin{bmatrix} x_1^T \\ x_2^T \\ \vdots \\ x_M^T \end{bmatrix}
= x_1x_1^T + x_2x_2^T + \cdots + x_Mx_M^T
= \sum_{i=1}^M x_ix_i^T
\end{align}
\end{split}

이러한 행렬들을 **로우-랭크 행렬(low-rank matrix)**라고 한다.

```python
# 연습문제

# x1 = [1, 1]
# 위 벡터로 랭크-1 행렬을 만들고 rank값 확인

x1 = np.array([[1],[1]])
np.linalg.matrix_rank(x1@x1.T)
```

```python
# x1 = [1,1]
# x2 = [1, -1]
# 위 벡터를 이용해 rank-2 행렬을 만들고 rank 확인

x1 = np.array([[1],[1]])
x2 = np.array([[1],[-1]])

np.linalg.matrix_rank(x1 @ x1.T + x2 @ x2.T)
```

이렇게 정방행렬이 풀랭크인 경우 역행렬이 존재한다. 따라서 풀랭크의 여부는 역행렬의 존재 여부라 할 수 있으며 풀랭크인 경우 행렬식의 계산이 필요없다고 할 수 있다.

```python
# 행렬식의 값이 0이면 역행력이 존재 X

np.linalg.det(x1 @ x1.T + x2 @ x2.T)
```

## 벡터공간과 기저벡터

여러 벡터를 선형조합하면 수많은 벡터들을 만들 수 있고 벡터 N개가 서로 선형독립하면 이 벡터들을 선형조합해 만들 수 있는 수많은 벡터의 집합을 **벡터공간(vector space) V**라 하며 벡터공간의 차원을 $N$이라 한다. 그리고 그 벡터 공간을 만드는 기준이 되는 벡터들을 **기저벡터**라 한다.

\begin{align}
V = \{ c_1x_1 + \cdots + c_Nx_N \; \vert \; c_1, \ldots, c_N \in \mathbf{R} \} 
\end{align}

여기서 중요한 점은 벡터공간의 차원은 벡터의 차원이 아니라 기저벡터의 개수라는 점이다.

N차원 벡터 N개가 선형독립이면 이를 선형조합하여 모든 N차원 벡터를 만들 수 있다.

## 랭크와 역행렬

정방행렬의 랭크와 역행렬 사이에는 다음과 같은 정의가 성립한다.

**정방행렬이 풀랭크면 역행렬이 존재한다.**

**역행렬이 존재하면 풀랭크다.**

증명

정방행렬이 풀랭크면 정방행렬을 이루는 벡터들이 선형독립이라는 의미이다. 
따라서 기저벡터가 될 수 있으므로 차원만 맞다면 모든 N차원 벡터를 만들 수 있다.

아래의 식에서 $e$는 만들고자 하는 벡터이다.

\begin{split}
\begin{align}
Xc_1 = e_1 = \begin{bmatrix} 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix}
\end{align}
\end{split}

\begin{split}
\begin{align}
Xc_2 = e_2 = \begin{bmatrix} 0 \\ 1 \\ \vdots \\ 0 \end{bmatrix}
\end{align}
\end{split}

위 식들을 모으면 아래와 같다.

\begin{align}
X \begin{bmatrix} c_1 & c_2 & \cdots & c_N \end{bmatrix} = XC = I
\end{align}

즉, 정방행렬의 경우 $XC = I$이면 $CX = I$가 성립한다.

따라서

$$XC = CX = I$$

가 되므로 위 식을 따르는 행렬 C가 존재한다는 것을 알 수 있다. 즉 C는 역행렬임을 알 수 있다.

$$C = X^{-1}$$

## 벡터공간 투영

M개의 N차원 기저벡터 $v1, v2, ..., v_M$이 존재한다고 가정하자

여기서 M은 N보다 작고 모든 N차원 벡터 $x$에 대해 기저벡터 $v_1, v_2, ..., v_M$을 선형조합해 만든 벡터 $x^{||v}$를 **벡터 공간에 대한 투영벡터**, 차이 벡터 $x - x^{\Vert v} = x^{\perp v}$를 **벡터공간에 대한 직교벡터**라 한다.

\begin{align}
(x - x^{\Vert V}) \perp \{ v_1, v_2, \cdots, v_M \} 
\end{align}

## 정규직교인 기저벡터로 이루어진 벡터공간

기저벡터 $v_1, v_2, ..., v_M$이 정규직교이면 투영벡터 $x^{||v}$는 각 기저벡터에 대한 내적값으로 표현된다.

\begin{align}
x^{\Vert V} = (x^Tv_1)v_1 + (x^Tv_2)v_2 + \cdots + (x^Tv_M)v_M
\end{align}

투영벡터의 길이의 제곱은 각 기저벡터와의 내적의 제곱합이다.

\begin{align}
\Vert x^{\Vert V} \Vert^2 = \sum_{i=1}^M (x^Tv_i)^2 
\end{align}

벡터 $x$에서 벡터 $x^{||v}$를 뺀 직교벡터$x^\perp$가 기저벡터에 모두 직교한다는 것은 아래와 같이 증명한다.

\begin{split}
\begin{align}
\begin{aligned}
v_i^T(x - x^{\Vert v})
&= 
v_i^Tx -
v_i^T \left( (x^Tv_1)v_1 + (x^Tv_2)v_2 + \cdots + (x^Tv_M)v_M  \right) \\
&=
v_i^Tx - \left( (x^Tv_1)v_i^Tv_1 + (x^Tv_2)v_i^Tv_2 + \cdots + (x^Tv_M)v_i^Tv_M \right) \\
&=
v_i^Tx - x^Tv_i \\
&= 0
\end{aligned}
\end{align}
\end{split}

따라서 직교벡터 x⊥V는 기저벡터 v1,v2,⋯,vM으로 이루어진 벡터공간의 모든 벡터에 대해 직교한다.

이 사실로부터 벡터 x의 투영벡터 x∥V은 기저벡터 v1,v2,⋯,vM으로 이루어진 벡터공간의 모든 벡터 중에서 가장 벡터 x와 가까운 벡터라는 것도 알 수 있다.

## 표준기저벡터

기저벡터 중에서도 원소 중 하나의 값만이 1이고 다른 값들은 모두 0으로 이루어진 기저벡터를 **표준기저벡터**라고 한다.

\begin{split} 
\begin{align}
e_1 = \begin{bmatrix} 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix}, \;\;
e_2 = \begin{bmatrix} 0 \\ 1 \\ \vdots \\ 0 \end{bmatrix}, \;\; 
\cdots,  \;\;
e_N = \begin{bmatrix} 0 \\ 0 \\ \vdots \\ 1 \end{bmatrix}
\end{align}
\end{split}

위의 벡터들을 행렬로 만들면 항등행렬이 된다.

\begin{align}
\begin{bmatrix} e_1 & e_2 & \cdots & e_N \end{bmatrix} = I_N
\end{align}

## 좌표

어떤 벡터의 **좌표**는 기저벡터를 선형조합하여 그 벡터를 나타내기 위한 계수벡터를 말한다.

예를 들어 다음과 같은 기저벡터 {$e_1, e_2$}를 선형조합하여 벡터 $x$를 나타낼 수 있다고 가정하자.

\begin{align}
x = x_{e_1} e_1 + x_{e_2} e_2  
\end{align}

이 때 벡터 $x_e$는

\begin{split} 
\begin{align}
x_e = \begin{bmatrix} x_{e_1} \\ x_{e_2} \end{bmatrix}  
\end{align}
\end{split}

이를 벡터 $x$의 좌표벡터, 좌표라고 한다.

만약 원점이 0,0 이라면 $e_1, e_2$는 단위벡터 (0,1), (1,0)일 것이다.

\begin{split} 
\begin{align}
x = [ e_1 e_2 ] \begin{bmatrix} x_{e_1} \\ x_{e_2} \end{bmatrix} = [ e_1 e_2 ] \; x_e 
\end{align}
\end{split}

표준기저벡터를 모아놓은 행렬은 앞서 항등행렬인 것을 확인했다. 따라서 표준기저벡터에 대한 벡터의 좌표는 원래 벡터와 같다. 하지만 같은 벡터라도 다른 기저벡터를 사용하면 그 좌표가 달라지므로 하나의 기저벡터에 따라 여러 좌표를 가질 수 있다.

```python
import matplotlib.pylab as plt

gray = {"facecolor": "gray"}
black = {"facecolor": "black"}
red = {"facecolor": "red"}
green = {"facecolor": "green"}
blue = {"facecolor": "blue"}
lightgreen = {"facecolor": "lightgreen"}
```

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])
x = np.array([2, 2])

plt.annotate('', xy=2 * e1, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=2 * e2, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=e1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=e2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=x, xytext=(0, 0), arrowprops=gray)

plt.plot(0, 0, 'ro', ms=10)
plt.plot(x[0], x[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$x$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.grid()
plt.show()
```

다음 기저벡터 {$g_1, g_2$}를 사용하면,

\begin{split}
\begin{align}
g_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \; g_2 = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
\end{align}
\end{split}

위의 벡터 $x$를 다음과 같이 표현할 수 있다.

```python
g1 = np.array([1, 0])
g2 = np.array([-1, 1])
x = np.array([2, 2])

plt.annotate('', xy=4 * g1, xytext=(0, 0), arrowprops=lightgreen)
plt.annotate('', xy=2 * g2, xytext=(0, 0), arrowprops=lightgreen)
plt.annotate('', xy=g1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=g2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=x, xytext=(0, 0), arrowprops=gray)

plt.plot(0, 0, 'ro', ms=10)
plt.plot(x[0], x[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$x$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$g_2$", fontdict={"size": 18})
plt.text(0.5, -0.5, "$g_1$", fontdict={"size": 18})

plt.xticks(np.arange(-10, 10))
plt.yticks(np.arange(-10, 10))
plt.xlim(-3, 7)
plt.ylim(-2, 5)
plt.grid()
plt.show()
```

## 변환행렬

원래의 기저벡터가 아닌 새로운 기저벡터가 있다고 하자. 이 새로운 기저벡터들의 기존 기저벡터에 대한 좌표를 열벡터로 보고 이를 행렬로 묶은 행렬 A를 생각하자.

예를 들어, 기존의 기저벡터가 {e1,e2}이고 새로운 기저벡터 {g1,g2}간에 다음과 같은 관계가 성립한다면,

\begin{split}
\begin{align}
\begin{aligned}
g_1 &=  \dfrac{1}{\sqrt{2}} e_1 + \dfrac{1}{\sqrt{2}} e_2 \\
g_2 &= -\dfrac{1}{\sqrt{2}} e_1 + \dfrac{1}{\sqrt{2}} e_2
\end{aligned}
\end{align}
\end{split}

e1, e2에 대한 g1, g2의 좌표벡터는 다음처럼 열벡터로 나타낼 수 있다.

\begin{split}
\begin{align}
g_{1e} = 
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} 
\end{bmatrix}, \;\;
g_{2e} = 
\begin{bmatrix} 
-\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\end{align}
\end{split}

두 좌표벡터들을 합쳐서 행렬로 표시하면 다음과 같다.

\begin{align}
\begin{bmatrix} g_1 & g_2 \end{bmatrix}
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} \begin{bmatrix} g_{1e} & g_{2e} \end{bmatrix}
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} A
\end{align}

\begin{split} 
\begin{align}
A =
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & -\dfrac{1}{\sqrt{2}} \\
\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\end{align}
\end{split}

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])
a = np.array([2, 2])
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)

plt.annotate('', xy=e1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=e2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=g1, xytext=(0, 0), arrowprops=red)
plt.annotate('', xy=g2, xytext=(0, 0), arrowprops=red)

plt.text(-0.18, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})
plt.text(0.3, 0.5, "$g_1$", fontdict={"size": 18})
plt.text(-0.45, 0.2, "$g_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.2, 1.7)
plt.ylim(-0.5, 1.3)
plt.grid()
plt.show()
```

## 좌표변환

새로운 기저벡터에 대해 좌표를 계산하는 것을 **좌표변환**이라 한다.

2차원인 경우 벡터 $x$의 기저벡터 {$e_1, e_2$}에 대한 좌표 $x_e$를 새로운 기저벡터 {$g_1, g_2$}에 대한 좌표 $x_g$로 변환한다.

\begin{align}
x = x_{e1}e_1 + x_{e2}e_2 = x_{g1} g_1 + x_{g2} g_2  
\end{align}

\begin{align}
x 
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} x_e 
= \begin{bmatrix} g_1 & g_2 \end{bmatrix} x_g 
\end{align}

이 식에

\begin{align}
\begin{bmatrix} g_1 & g_2 \end{bmatrix}
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} 
A
\end{align}

를 대입하면

\begin{align}
x 
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} x_e 
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} A x_g
\end{align}

가 되므로 다음 식이 성립한다.

$$x_g = A^{-1}x_e = Tx_e$$

\begin{align}
x_g = A^{-1}x_e = Tx_e
\end{align}

위 식에서 $A$의 역행렬 $T = A^{-1}$을 **변환행렬**이라 한다.

예를 들어 벡터 $x$의 표준기저벡터에 대한 좌표가 다음과 같다고 가정하자.

\begin{split}
\begin{align}
x = 2 e_1 + 2 e_2= \begin{bmatrix}2 \\ 2\end{bmatrix} = x_e
\end{align}
\end{split}

여기서 표준 기저벡터에 대한 새로운 기저벡터의 좌표가 다음과 같다면

\begin{split} 
\begin{align}
g_{1e} = \begin{bmatrix} \dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix} ,\;\;
g_{2e} = \begin{bmatrix} -\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix}
\end{align}
\end{split}

새로운 기저벡터에 대한 $a$의 좌표는 다음과 같이 계산한다.

\begin{split} 
\begin{align}
x_g = A^{-1}x_e = 
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & -\dfrac{1}{\sqrt{2}} \\
\dfrac{1}{\sqrt{2}} &  \dfrac{1}{\sqrt{2}} 
\end{bmatrix}^{-1}
\begin{bmatrix}2 \\ 2\end{bmatrix}
=
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} \\
-\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\begin{bmatrix}2 \\ 2\end{bmatrix} 
= \begin{bmatrix}2\sqrt{2}\\0\end{bmatrix}
\end{align}
\end{split}

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])
x = np.array([2, 2])
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)

plt.annotate('', xy=e1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=e2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=x, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=g1, xytext=(0, 0), arrowprops=red)
plt.annotate('', xy=g2, xytext=(0, 0), arrowprops=red)

plt.plot(0, 0, 'ro', ms=10)
plt.plot(x[0], x[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$x$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})
plt.text(0.2, 0.5, "$g_1$", fontdict={"size": 18})
plt.text(-0.6, 0.2, "$g_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.grid()
plt.show()
```

즉, 새로운 좌표벡터는 원래의 좌표벡터에 변환행렬을 곱하여 구할 수 있다.

## 이미지 변환

새로운 기저벡터에 대한 좌표변환을 응용해 이미지를 자유롭게 변환할 수 있다.

```python
import scipy as sp
import scipy.misc
import scipy.ndimage

f = sp.misc.face(gray=True)

e1 = np.array([0, 1]) # matplotlib의 이미지 좌표규약으로 부호변경 
e2 = np.array([1, 0])
E = np.vstack([e1, e2]).T
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)
A = np.vstack([g1, g2]).T
gc1 = E @ g1
gc2 = E @ g2

plt.subplot(121)
plt.imshow(f, cmap=mpl.cm.bone, alpha=0.9)
plt.annotate('', xy=500*e1, xytext=(0,0), arrowprops=green)
plt.annotate('$e_1$', xy=500*e1, xytext=500*e1 + [-100,0])
plt.annotate('', xy=500*e2, xytext=(0,0), arrowprops=green)
plt.annotate('$e_2$', xy=500*e2, xytext=500*e2 + [0, -50])
plt.annotate('', xy=500*gc1, xytext=(0, 0), arrowprops=red)
plt.annotate('$g_1$', xy=500*gc1, xytext=500*gc1 + [50, -50])
plt.annotate('', xy=500*gc2, xytext=(0, 0), arrowprops=red)
plt.annotate('$g_2$', xy=500*gc2, xytext=500*gc2 + [50, 0])
plt.axis("off")
plt.xlim(-200, 1000)
plt.ylim(800, -500)
plt.title("좌표변환전")

f1 = sp.ndimage.affine_transform(f, A)

plt.subplot(122)
plt.imshow(f1, cmap=mpl.cm.bone, alpha=0.8)
plt.annotate('', xy=500*e1, xytext=(0,0), arrowprops=red)
plt.annotate('$g_1$', xy=500*e1, xytext=500*e1 + [-100,0])
plt.annotate('', xy=500*e2, xytext=(0,0), arrowprops=red)
plt.annotate('$g_2$', xy=500*e2, xytext=500*e2 + [0, -50])
plt.axis("off")
plt.xlim(-200, 1000)
plt.ylim(800, -500)
plt.title("좌표변환후")
plt.show()
```
