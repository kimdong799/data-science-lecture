## 1차원 근사

2차원 평면 위의 임의의 3개의 2차원 벡터 $a_1, a_2, a_3$가 있다고 가정하자. 벡터들을 어떤 마을이라 하고 원점을 수도라고 했을 때 수도에서 각 마을을 연결하는 가능한 한 가까이 있는 직선을 만들기 위해선 직선의 방향을 어떻게 설정해야할까? 이 직선의 방향을 나타내는 단위 벡터를 $w$라고 하자.

```python
w = np.array([2, 1]) / np.sqrt(5)
a1 = np.array([3, -1])
a2 = np.array([1, 3])
a3 = np.array([1, 1])

black = {"facecolor": "black"}

plt.figure(figsize=(9, 6))
plt.plot(0, 0, 'kP', ms=10)
plt.annotate('', xy=w, xytext=(0, 0), arrowprops=black)
plt.plot([-2, 8], [-1, 4], 'b--', lw=2)
plt.plot([a1[0], 2], [a1[1], 1], 'g:', lw=2)
plt.plot([a2[0], 2], [a2[1], 1], 'g:', lw=2)
plt.plot([a3[0], 1.2], [a3[1], 0.6], 'g:', lw=2)
plt.plot(a1[0], a1[1], 'ro', ms=10)
plt.plot(a2[0], a2[1], 'ro', ms=10)
plt.plot(a3[0], a3[1], 'ro', ms=10)
plt.text(0.1, 0.5, "$w$")
plt.text(a1[0] + 0.2, a1[1] + 0.2, "$a_1$")
plt.text(a2[0] + 0.2, a2[1] + 0.2, "$a_2$")
plt.text(a3[0] - 0.3, a3[1] + 0.2, "$a_3$")
plt.xticks(np.arange(-3, 15))
plt.yticks(np.arange(-1, 5))
plt.xlim(-3, 6)
plt.ylim(-2, 4)
plt.grid()
plt.show()
```

**$a_1, a_2, a_3$에서 직선을 내린 거리의 제곱합이 가장 작아지도록 하는 방향 $w$**를 찾는 것이 이 문제의 해답이 된다.

이러한 문제를 **1차원 근사문제 혹은 트롤리문제** 라고 한다.

벡터 $w$와 점 $a_i$의 거리의 제곱은 다음과 같이 계산한다.

\begin{align}
\Vert a_i^{\perp w}\Vert^2 = \Vert a_i\Vert^2 - \Vert a_i^{\Vert w}\Vert^2 = \Vert a_i\Vert^2 - (a_i^Tw)^2
\end{align}

즉, 벡터 $a$에 대해 투영성분의 놈을 뺀 값이 된다.

하나의 벡터에 대한 거리의 제곱은 위와 같이 계산한다. 그런데 이 문제의 경우 세개의 마을, 세개의 벡터가 존재했기에 다음과 같이 $a_1, a_2, a_3$를 행벡터로 가지는 행렬 $A$를 가정한다.

\begin{split} 
\begin{align}
A = \begin{bmatrix} a_1^T \\ a_2^T \\ a_3^T \end{bmatrix}
\end{align}
\end{split}

**행벡터의 놈의 제곱의 합은 행렬의 놈**이므로 모든 점들과의 거리의 제곱의 합은 행렬의 놈으로 계산된다.

\begin{split} 
\begin{align}
\begin{aligned}
\sum_{i=1}^3 \Vert a_i^{\perp w}\Vert^2 
&= \sum_{i=1}^3 \Vert a_i\Vert^2 - \sum_{i=1}^3  (a_i^Tw)^2 \\
&= \Vert A \Vert^2 - \Vert Aw\Vert^2 \\
\end{aligned}
\end{align}
\end{split}

여기서 추가적인 조건은 우리가 찾고자 하는 방향 $w$가 이 세 벡터와의 거리의 제곱합이 최소가 되는 지점이다. 

그런데 점 $a_i$의 위치는 고정되어 있기에 행렬 $A$의 놈 값은 고정되어있다. 따라서 $||Aw||^2$ 값이 가장 커지게 하는 벡터 $w$를 찾아야 한다.

\begin{align}
\arg\max_w \Vert Aw \Vert^2 
\end{align}

## 1차원 근사의 풀이

1차원 근사 문제를 다시 한 번 살펴보자.

```python
w = np.array([2, 1]) / np.sqrt(5)
a1 = np.array([3, -1])
a2 = np.array([1, 3])
a3 = np.array([1, 1])

black = {"facecolor": "black"}

plt.figure(figsize=(9, 6))
plt.plot(0, 0, 'kP', ms=10)
plt.annotate('', xy=w, xytext=(0, 0), arrowprops=black)
plt.plot([-2, 8], [-1, 4], 'b--', lw=2)
plt.plot([a1[0], 2], [a1[1], 1], 'g:', lw=2)
plt.plot([a2[0], 2], [a2[1], 1], 'g:', lw=2)
plt.plot([a3[0], 1.2], [a3[1], 0.6], 'g:', lw=2)
plt.plot(a1[0], a1[1], 'ro', ms=10)
plt.plot(a2[0], a2[1], 'ro', ms=10)
plt.plot(a3[0], a3[1], 'ro', ms=10)
plt.text(0.1, 0.5, "$w$")
plt.text(a1[0] + 0.2, a1[1] + 0.2, "$a_1$")
plt.text(a2[0] + 0.2, a2[1] + 0.2, "$a_2$")
plt.text(a3[0] - 0.3, a3[1] + 0.2, "$a_3$")
plt.xticks(np.arange(-3, 15))
plt.yticks(np.arange(-1, 5))
plt.xlim(-3, 6)
plt.ylim(-2, 4)
plt.grid()
plt.show()
```

1차원 근사 문제는 위와 같은 2차원 좌표 평면도 상에 점 $a_1, a_2, a_3$에서 수직하면서 각 데이터 포인트의 거리의 제곱합이 가장 작아지는 w벡터의 방향을 찾는 문제이다.

여기서 점 $a_1, a_2, a_3$를 어떤 마을이라고 했을 때 벡터 $w$의 방향으로 이동한 데이터 포인트를 각 마을에서 가장 가까운 역이라고 할 수 있다. 이처럼 **1차원 근사 문제는 2차원 상에 있는 데이터 포인트를 1차원(직선)상에 근사화 하는 작업**인 것이다.

이는 이후 학습할 PCA의 기본 원리가 된다.

앞의 1차원 근사 문제의 해결 방법에서 $A$는 원래 데이터의 위치이고 $w$는 근사화 할 방향이라고 설명했다.

따라서 다음과 같이 계산된다.

$$Aw$$

그런데 여기서 $A$는 이미 고정된 값이므로 $w$를 조정해야한다.

따라서

$$||Aw||$$가 가능한 커지게 하는 $w$를 찾는 문제가 된다.

$$w^* = arg max_w||Aw||$$

**$w$는 방향을 나타내는 단위벡터이므로 $||w|| = 1 이다.$**

여기서 $w$를 찾기 위해 특이분해를 이용한다.

위에서 예시로 든 행렬

$$A \in \mathbf{R}^{3 \times 2}$$

를 특이분해하면 다음과 같은 모양이 된다.

여기서 왼쪽 특이벡터 중 $u_3$는 사실상 의미가 없으므로 축소형으로 표현하면 다음과 같아진다.

최종적으로 $A$ 행렬을 특이분해하면 다음과 같이 특이분해된다.

- 첫 번째 특잇값 : $\sigma_1$, 첫번째 왼쪽 특이벡터 $u_1$, 첫번째 오른쪽 특이벡터 $v_1$
- 두 번째 특잇값 : $\sigma_2$, 두번째 왼쪽 특이벡터 $u_2$, 두번째 오른쪽 특이벡터 $v_2$

여기서 특잇값의 성질에 따라 다음과 같은 정리가 성립된다.

1. 특잇값은 내림차순으로 정렬되므로 첫번째 특이값 $\sigma_1$은 두 번째 특이값 $\sigma_2$보다 같거나 크다.

\begin{align}
\sigma_1 \geq \sigma_2
\end{align}

2. $Av = \sigma u$가 성립하므로

$$Av_1 = \sigma_1 u_1$$
$$Av_2 = \sigma_2 u_2$$

$A$에 오른쪽 특이벡터를 곱하면 왼쪽 특이벡터 방향이 된다.

3. 오른쪽 특이벡터 $V$는 서로 직교하는 성질을 가진다. 따라서 서로 독립하는 행렬이기 때문에 기저벡터가 될 수 있다.

서로 독립하는 기저벡터 $V = $ {$V_1, V_2$}

따라서 2차원 기저벡터 $V$가 존재하므로 모든 2차원 상의 벡터를 표현할 수 있다.

그런데 1차원 근사문제에서 찾고자 하는 $w$ 역시 2차원 벡터이다.

따라서 $w$는 $v_1, v_2$ 벡터를 잘 선형조합하여 찾을 수 있다.

$$w = w_1v_1 + w_2v_2$$

결국 1차원 근사문제는 기저벡터 $v_1, v_2$에 대해 선형조합되는 $w_1, w_2$를 찾는 문제로 바뀌게 된다.

**조건!! $w$ 벡터는 방향만을 나타내는 단위벡터로 그 길이는 1이다.**

$$w^2_1 + w^2_2 = 1$$

1차원 근사 문제의 목적은 $||Aw||$의 값을 가장 크게 만드는 것이다.

이 때 $$||Aw||$$의 값은 다음과 같다.

$w$는 $v$의 선형조합으로 표현 가능하므로

$||Aw||^2 = ||A(w_1v_1 + w_2v_2)||^2$

위 식에서 $w_1, w_2$값은 스칼라 값으로 앞으로 빼주면

$= ||w_1Av_1 + w_2Av_2||^2$

위 식에서 $Av$는 $\sigma u$로 표현 가능하므로

$= ||w_1 \sigma_1 u_1 + w_2 \sigma_2 u_2||^2$

위 식에서 $u$는 직교하므로

$= ||w_1 \sigma_1 u_1||^2 + ||w_2 \sigma_2 u_2||^2$ (직교)

위 식에서 $w, \sigma$는 스칼라 값으로 앞으로 빼주면

$= w^2_1\sigma^2_1 ||u_1||^2 + w^2_2\sigma^2_2 ||u_2||^2$

가 된다. 여기서 $u$는 단위벡터이므로 놈 값은 1이된다.

따라서

$= w^2_1\sigma_1^2 + w_2^2 \sigma_2^2(단위벡터)$

가 된다.

**최종적으로 1차원 근사문제에서 $Aw$의 값을 가장 크게 만드는 것은 아래의 값을 가장 크게 만드는 문제가 된다.**

$w^2_1\sigma_1^2 + w_2^2 \sigma_2^2$

그렇다면 $w_1, w_2$를 어떤 값으로 지정해야 할까?

$Aw$가 가장 크게 만드는 $w$를 찾는 문제이기 때문에 단순히 100, 1000 을 곱하면 될까?

정답은 아니다. $w$는 방향을 나타내는 단위벡터로 $w_1^2 + w_2^2$의 값이 1이 된다는 조건 하에서 $Aw$를 최대로 만드는 값을 찾아야 한다.

이 문제를 해결하기 위해 하나의 예시를 이용해보자.

가장 많은 돈을 벌어야 하는 상황에서 금 1Kg, 쇠 1Kg이 있을 때 무엇이든 총합 1Kg만을 가져올 수 있다면 금만 1Kg을 가져올 것이다.

그렇다면 아래의 식에서 $\sigma$값이 가장 큰 항에 가중치를 1로 두면 되지 않을까?

$w^2_1\sigma_1^2 + w_2^2 \sigma_2^2$

위 식을 잘 살펴보면 다음의 정리를 떠올릴 수 있다.

$\Sigma$는 단위행렬이면서 $\sigma$값이 큰것부터 작은것 순서로 단위행렬이 구성된다.

그렇다면 $\sigma_1$은 $\sigma_2$보다 항상 크다는 논리가 성립한다.  따라서 다음과 같이 가중치를 둘 수 있다.

$$w_1 = 1, w_2 = 0$$

즉, 찾고자 하는 값은 첫번째 오른쪽 특이벡터 방향과 같아진다.

$||Aw||^2 = ||A(w_1v_1 + w_2v_2)||^2$

의 식에서 $w_1$은 1이므로

$$w = v_1$$

이 때, $||Aw||는 첫 번째 특잇값이 된다.$

$$||Aw|| = ||Av_1|| = ||\sigma_1 u_1|| = \sigma_1 ||u_1|| = \sigma_1$$
