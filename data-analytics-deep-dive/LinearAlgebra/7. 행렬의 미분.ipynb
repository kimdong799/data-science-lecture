## 행렬의 미분

기본적인 함수는 스칼라 값을 받아 스칼라 값을 출력하는 함수이다. 하지만 여러개의 입력을 가지는 **다변수함수**는 함수의 독립변수가 벡터인 경우라고 할 수 있다.

여러개의 입력을 받는 다변수함수를 생각해보자.

다변수함수는 여러개의 입력을 받고 다음과 같이 벡터를 입력으로 받는다고 할 수 있다.

\begin{split}
\begin{align}
f\left( \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \right) = f(x) = f(x_1, x_2)
\end{align}
\end{split}

$$벡터 x \rightarrow 스칼라 f$$

이 개념을 확장하면 행렬을 입력으로 가지는 함수를 표현할 수 있다.

\begin{split} 
\begin{align}
f\left( \begin{bmatrix} x_{11} & x_{12} \\ x_{21} & x_{22} \end{bmatrix} \right) 
= f(X) = f(x_{11}, \cdots, x_{22})
\end{align}
\end{split}

$$행렬 X \rightarrow 스칼라f$$

반대로 벡터나 행렬을 출력하는 함수는 여러 함수를 합쳐놓은 것이라고 할 수 있다.

\begin{split} 
\begin{align}
f(x) =
\begin{bmatrix}
f_1(x) \\
f_2(x)
\end{bmatrix}
\end{align}
\end{split}

$$스칼라 x \rightarrow 벡터 f$$

마찬가지로 개념을 확장해 행렬을 출력으로 본다면

\begin{split} 
\begin{align}
f(x) =
\begin{bmatrix}
f_{11}(x) & f_{12}(x) \\
f_{21}(x) & f_{22}(x)
\end{bmatrix}
\end{align}
\end{split}

$$스칼라 f \rightarrow 행렬 f$$

마지막으로 행렬을 입력받아 벡터나 행렬을 출력하는 경우이다.

\begin{split} 
\begin{align}
f(x) =
\begin{bmatrix}
f_1(x_1, x_2) \\
f_2(x_1, x_2)
\end{bmatrix}
\end{align}
\end{split}

$$벡터 x \rightarrow 벡터 f$$

\begin{split} 
\begin{align}
f(x) =
\begin{bmatrix}
f_{11}(x_1, x_2) & f_{12}(x_1, x_2) \\
f_{21}(x_1, x_2) & f_{22}(x_1, x_2)
\end{bmatrix}
\end{align}
\end{split}

$$벡터 x \rightarrow 행렬 f$$

이렇게 행렬을 입력이나 출력으로 갖는 함수를 미분하는 행위를 **행렬미분**이라 한다.

## 스칼라를 벡터로 미분하는 경우

데이터분석에서는 입력변수 $x$가 벡터이고 출력변수가 스칼라 $f$인 다변수함수를 사용하는 경우가 많다.

따라서 편미분도 $\frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \cdots$와 같이 다양하게 존재한다.

이처럼 스칼라값을 벡터로 미분하는 경우 결과를 열벡터로 표현하며 이렇게 만들어진 벡터를 **그래디언트 벡터**라고 하며 $\nabla f$로 표기한다.

\begin{split}
\begin{align}
\nabla f = 
\frac{\partial f}{\partial {x}} =
\begin{bmatrix}
\dfrac{\partial f}{\partial x_1}\\
\dfrac{\partial f}{\partial x_2}\\
\vdots\\
\dfrac{\partial f}{\partial x_N}\\
\end{bmatrix}
\end{align}
\end{split}

위 식에서 $x$는 vector를 의미하며 스칼라$f$를 벡터 $x$로 미분한 결과이다.

따라서 기울기 벡터의 갯수는 $x$의 갯수만큼 존재한다.

### 연습문제

다음의 함수에서 그래디언트 벡터를 구하라.

다음 그래디언트 벡터에 대해 x, y가 다음 위치일 때 그래디언트 벡터의 값을 구하고 평면상에 화살표로 표현하라.

2차원의 경우 그래디언트 벡터를 표시할 수 있는데 2개의 입력변수를 가지는 2차원 함수 $f(x, y)$는 평면상에 **컨투어플롯**으로 나타낼 수 있다.

$x, y$위치에서의 그래디언트 벡터는 그 위치를 원점으로 하는 화살표로 표현할 수 있다.

그래디언트 벡터의 방향은 편미분 성분 $\frac{\partial f}{\partial x}$와 $\frac{\partial f}{\partial y}$의 부호에 의해 결정된다.

만약 어떤 위치 $x, y$에서 $x$가 증가할수록 $f$가 증가하면 도함수 $\frac{\partial f}{\partial x}$는 양수이다.

만약 어떤 위치 $x, y$에서 $y$가 증가할수록 $f$가 감소하면 도함수 $\frac{\partial f}{\partial x}$는 음수이다.

이 컨투어 플롯 위에 그래디언트 벡터를 화살표로 표현한 플롯을 퀴버플롯이라고 한다.

퀴버플롯에서 화살표는 화살표 시작 지점의 그래디언트 벡터를 나타낸다.

## 퀴버플롯

다음의 함수에서의 그래디언트 벡터를 표시한 퀴버플롯을 살펴보자.

\begin{align}
2x^2 + 6xy + 7y^2 - 26x - 54y + 107
\end{align}

```python
def f(x, y):
    return 2 * x**2 + 6 * x * y + 7 * y**2 - 26 * x - 54 * y + 107

xx = np.linspace(1, 16, 100)
yy = np.linspace(-3, 6, 90)
X, Y = np.meshgrid(xx, yy)
Z = f(X, Y)


def gx(x, y):
    return 4 * x + 6 * y - 26

def gy(x, y):
    return 6 * x + 14 * y - 54

xx2 = np.linspace(1, 16, 15)
yy2 = np.linspace(-3, 6, 9)
X2, Y2 = np.meshgrid(xx2, yy2)
GX = gx(X2, Y2)
GY = gy(X2, Y2)
plt.figure(figsize=(10, 5))
plt.contour(X, Y, Z, levels=np.logspace(0, 3, 10))
plt.quiver(X2, Y2, GX, GY, color='blue', scale=400, minshaft=2)

plt.xlabel('x')
plt.ylabel('y')
plt.title("퀴버 플롯(quiver plot)")
plt.grid()
plt.show()
```

**퀴버플롯에서의 그래디언트 벡터의 특징**

1. 그래디언트 벡터의 크기는 기울기. 즉, 벡터의 크기가 클수록 함수 곡면의 기울기가 커진다.
2. 그래디언트 벡터의 방향은 함수 곡면의 기울기가 가장 큰 방향, 즉 단위 길이단 함숫값이 가장 크게 증가하는 방향
3. 그래디언테 벡터의 방향은 등고선 방향과 직교

위 특징을 따라 위의 퀴버플롯은 가운데가 움푹 파인 오목한 그릇(bowl)모양인 것을 유추할 수 있다.

## 행렬미분법칙

선형모형을 미분하면 **그래디언트 벡터는 가중치 벡터이다.**

\begin{align}
f(x) = w^T x 
\end{align}

\begin{align}
\nabla f = \frac{\partial {w}^{T}{x}}{\partial {x}} = \frac{\partial {x}^{T}{w}}{\partial {x}} = {w} 
\end{align}

이차형식을 미분하면 **행렬과 벡터의 곱으로 나타난다.**

\begin{align}
f(x) = x^T A x
\end{align}

\begin{align}
\nabla f(x) = \frac{\partial {x}^{T}{A}{x}}{\partial {x}} = ({A} + {A}^{T}){x}
\end{align}

## 벡터를 스칼라로 미분하는 경우

\begin{split}
\begin{align}
{f}(x) =
\begin{bmatrix}
f_1 \\
f_2 \\
\vdots\\
f_M \\
\end{bmatrix}
\end{align}
\end{split}

위의 벡터를 스칼라 $x$로 미분하는 경우 결과는 행벡터로 표시한다.

\begin{align}
\frac{\partial {f}}{\partial x} = 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial x} &
\dfrac{\partial f_2}{\partial x} &
\cdots &
\dfrac{\partial f_M}{\partial x}
\end{bmatrix}
\end{align}

## 벡터를 벡터로 미분하는 경우

벡터를 벡터로 미분하는 경우는 벡터 $x$를 입력으로 받아 벡터$f$를 출력하는 함수이다.

즉, 도함수는 2차원 배열, 행렬이 된다.

\begin{split}
\begin{align}
\dfrac{\partial {f}}{\partial {x}}
= 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial {x}} &
\dfrac{\partial f_2}{\partial {x}} &
\cdots                             & 
\dfrac{\partial f_N}{\partial {x}}
\end{bmatrix}
=
\begin{bmatrix}
\dfrac{\partial {f}}{\partial x_1} \\
\dfrac{\partial {f}}{\partial x_2} \\
\vdots                             \\
\dfrac{\partial {f}}{\partial x_M}
\end{bmatrix}
= 
\begin{bmatrix}
\dfrac{\partial f_1}{\partial x_1} & \dfrac{\partial f_2}{\partial x_1} & \cdots & \dfrac{\partial f_N}{\partial x_1} \\
\dfrac{\partial f_1}{\partial x_2} & \dfrac{\partial f_2}{\partial x_2} & \cdots & \dfrac{\partial f_N}{\partial x_2} \\
\vdots & \vdots & \ddots & \vdots \\
\dfrac{\partial f_1}{\partial x_M} & \dfrac{\partial f_2}{\partial x_M} & \cdots & \dfrac{\partial f_N}{\partial x_M} \\
\end{bmatrix}
\end{align}
\end{split}

## 행렬과 벡터의 곱의 미분

행렬 $A$와 벡터 $x$의 곱 $Ax$를 벡터 $x$로 미분하면 행렬 $A^T$가 된다.

\begin{align}
\nabla f(x) = \dfrac{\partial ({Ax})}{\partial {x}} = A^T 
\end{align}

함수의 출력변수와 입력변수가 모두 벡터(다차원)인 경우 입력변수, 출력변수에 대한 모든 조합에 대해 미분이 존재한다.

따라서 도함수는 행렬 형태가 된다.

이렇게 만들어진 행렬을 **자코비안 행렬**이라고 하며 벡터함수를 벡터변수로 미분해서 생기는 행렬의 전치행렬이다.

쉽게 생각하면 자코비안 행렬은 그래디언트 행렬이라고 생각할 수 있다.

### 헤시안 행렬

다변수함수의 2차 도함수는 그래디언트 벡터를 입력변수 벡터로 미분한 것으로 헤시안 행렬이라고 한다.

헤시안 행렬은 그래디언트 벡터의 자코비안 행렬의 전치행렬로 정의한다.

## 스칼라를 행렬로 미분

도함수 행렬의 모양이 입력변수 행렬 $x$와 같다.

\begin{split}
\begin{align}
\dfrac{\partial f}{\partial {X}} =
\begin{bmatrix}
\dfrac{\partial f}{\partial x_{1,1}} & \dfrac{\partial f}{\partial x_{1,2}} & \cdots & \dfrac{\partial f}{\partial x_{1,N}}\\
\dfrac{\partial f}{\partial x_{2,1}} & \dfrac{\partial f}{\partial x_{2,2}} & \cdots & \dfrac{\partial f}{\partial x_{2,N}}\\
\vdots & \vdots & \ddots & \vdots\\
\dfrac{\partial f}{\partial x_{M,1}} & \dfrac{\partial f}{\partial x_{M,2}} & \cdots & \dfrac{\partial f}{\partial x_{M,N}}\\
\end{bmatrix}
\end{align}
\end{split}

## 행렬 곱의 대각성분

두 정방행렬을 곱해서 만들어진 행렬의 대각성분은 스칼라이다.

스칼라를 뒤의 행렬로 미분하면 앞의 행렬의 전치행렬이다.

\begin{align}
f(X) = \text{tr} ({W}{X})
\end{align}

\begin{align}
\dfrac{\partial f}{\partial X} =
\dfrac{\partial \, \text{tr} ({W}{X})}{\partial {X}} = {W}^T
\end{align}

## 행렬식의 로그

행렬식은 스칼라이고 이 값의 로그값 또한 스칼라이다. 

이 값을 원래의 행렬로 미분하면 원래 행렬의 역행렬의 전치행렬이 된다

\begin{align}
f(X) = \log | {X} |
\end{align}

\begin{align}
\dfrac{\partial f}{\partial X} = \dfrac{\partial \log | {X} | }{\partial {X}} = ({X}^{-1})^T 
\end{align}
