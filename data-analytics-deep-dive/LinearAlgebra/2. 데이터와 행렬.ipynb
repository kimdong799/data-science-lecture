## 이미지 데이터 : MNIST

```python
from sklearn.datasets import load_digits

# 손글씨 이미지 데이터 로드
digits = load_digits()
samples = [0, 10, 20, 30, 1, 11, 21, 31] # 0과 1의 이미지 데이터 번호
d = []
for i in range(8):
    d.append(digits.images[samples[i]])

# 이미지 시각화
plt.figure(figsize=(8, 2))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title('image {}'.format(i + 1))
plt.suptitle('숫자 0과 1 이미지')
plt.tight_layout()
plt.show()
```

```python
# 이미지 데이터의 shape (해상도)

d[0].shape
```

위의 이미지를 벡터화시키기 위해 8*8 해상도 데이터를 64 * 1 shape의 벡터로 벡터화 진행

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64, 1)) # 8*8 해상도 데이터이기 때문에 64*1 shape으로 벡터화

plt.figure(figsize = (8,3))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(v[i], aspect = 0.4,
              interpolation = 'nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title('벡터 {}'.format(i + 1))
plt.suptitle('벡터화 된 이미지', y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

위 시각화 결과를 통해 인간은 각 벡터가 무슨 숫자인지 인지하기 어렵다.

하지만 벡터화된 결과를 통해 각각의 벡터에 대한 유사한 정도를 인지할 수 있다.

## 텐서

텐서란 같은 크기의 행렬이 여러 개 묶여 있는 것으로 예를 들어 컬러 이미지의 경우 해상도를 나타내는 2차원 행렬과 색상을 표현하기 위한 RGB 채널에 대한 3차원이 추가적으로 존재한다.

따라서 이미지 데이터의 해상도가 768 * 1021 일 때 이 컬러 이미지의 shape은 768 * 1024 * 3인 텐서이다.

```python
from scipy import misc
img_rgb = misc.face()
img_rgb.shape
```

```python
plt.subplot(221)
plt.imshow(img_rgb, cmap=plt.cm.gray)  # 컬러 이미지 출력
plt.axis("off")
plt.title("RGB 컬러 이미지")

plt.subplot(222)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.gray)  # red 채널 출력
plt.axis("off")
plt.title("Red 채널")

plt.subplot(223)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.gray)  # green 채널 출력
plt.axis("off")
plt.title("Green 채널")

plt.subplot(224)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.gray)  # blue 채널 출력
plt.axis("off")
plt.title("Blue 채널")

plt.show()
```

## 특수한 벡터와 행렬

### 영벡터(zeros-vector)


모든 원소가 0인 N차원 벡터를 영벡터라 한다.

```python
# 5차원 0벡터

np.zeros([5,1])
```

### 일벡터(ones-vector)

모든 원소가 1인 N차원 벡터를 일벡터라 한다.

```python
# 5차원 일벡터

np.ones([5, 1])
```

### 정방행렬(square matrix)

행과 열의 개수가 같은 행렬을 정방행렬이라고 한다.

```python
np.zeros([5,5])
```

### 대각행렬(diagonal matrix)

행과 열이 같은 위치를 주 대각(main diagonal) 또는 대각(diagonal)이라 한다. 

반대로 대각 위치에 있지 않은 것들은 비대각(off-diagonal)이라 한다. 이 때 모든 비대각 요소가 0인 행렬을 대각 행렬이라 한다.

반드시 정방행렬일 필요는 없으며 대각 성분이 0이 아니기만 하면 된다.

```python
np.diag([1, 2, 3])
```

### 항등행렬(identity matrix)

대각행렬 중 모든 대각 성분의 값이 1인 행렬을 항등행렬이라 하며 대문자 $I$로 표기한다.

```python
np.identity(5)
```

```python
np.eye(5)
```

### 대칭행렬(symmetric matrix)

전치연산의 결과인 전치행렬이 원래의 행렬과 같은 경우 이를 대칭행렬이라 한다. 즉, 정방행렬만이 대칭행렬이 될 수 있다.

```python
np.ones([5, 5])
```

```python
np.ones([5, 5]).T
```
