## 다항식함수

**다항식함수(polynomial)**는 상수항 $c_0, 일차항 c_1x, 이차항 c_2x^2, ...$등 거듭제곱 항의 선형조합으로 이루어진 함수이다.

$f(x) = c_0 + c_1x + c_2x^2 + ... + c_nx^n$

## 최대함수와 최소함수

최대함수는 두 인수 중 큰 값을 출력한다.

\begin{split} 
\begin{align}
\max(x, y) 
=
\begin{cases}
x & \text{ if } x \geq y \\
y & \text{ if } x < y 
\end{cases}
\end{align}
\end{split}

반대로 최소함수는 두 인수 중 작은 값을 출력한다.

\begin{split} 
\begin{align}
\min(x, y) 
=
\begin{cases}
x & \text{ if } x \leq y \\
y & \text{ if } x > y 
\end{cases}
\end{align}
\end{split}

하지만 실제 데이터 분석에서 최대함수와 최소함수를 위처럼 사용하는 경우는 거의 없고 일반적으로 $y=0$으로 고정시킨 후 입력값 $x$가 양수면 그대로 $x$를 출력하고 음수일 때는 0으로 만들 때 주로 사용한다.

\begin{split} 
\begin{align}
\max(x, 0) 
=
\begin{cases}
x & \text{ if } x \geq 0 \\
0 & \text{ if } x < 0 
\end{cases}
\end{align}
\end{split}

이후 학습할 인공신경망에서 이 함수를 **ReLU(Rectified Linear Unit)**라고 부른다.

```python
np.maximum(xx, 0)
```

```python
xx = np.linspace(-10, 10, 100)
plt.plot(xx, np.maximum(xx, 0))
plt.title("max(x,0) 또는 ReLU")
plt.xlabel("$x$")
plt.ylabel("$ReLU(x)$")
plt.grid()
plt.show()
```

## 지수함수

정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수가 있다고 가정하자.

$y = a^x$

위 식에서 거듭제곱되는 승수를 입력값으로 하는 함수가 바로 지수함수이다.

위 식에서 $a$를 밑(base)이라고 하며 데이터 분석에서는 주로 **오일러 수 $e$(약 2.718)로 거듭제곱 하는 함수를 지수함수(exponential function)**라고 한다.

$y = e^x$

혹은

$y = exp(x) = exp$ $x$

```python
# 오일러 수 e

np.e
```

```python
np.exp(-10), np.exp(-1), np.exp(-0.1), np.exp(0), np.exp(0.1), np.exp(1), np.exp(10)
```

**지수함수의 특징**

1. 양수($e$)를 거듭제곱한 값으로 항상 양수이다.
2. $x=0$일 때 1이 된다. (모든 수의 0제곱은 1)
3. $x$가 무한대로 가면 양의 무한대로 다가간다.
4. $x$가 음의 무한대로 가면 0으로 다가간다.
5. 단조증가의 특징으로 $x_1 > x_2 이면 exp x_1 > exp x_2 이다.$

```python
xx = np.linspace(-2, 2, 100)
yy = np.exp(xx)
plt.title("지수함수")
plt.plot(xx, yy)
plt.axhline(1, c='r', ls="--")
plt.axhline(0, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.xlabel("$x$")
plt.ylabel("$\exp(x)$")
plt.show()
```

**지수함수의 가장 중요한 특징 중 하나**

두 지수함수의 곱의 값은 입력값 합의 지수함수와 같다.

\begin{align}
f(x_1) \cdot f(x_2) = e^{x_1} \cdot e^{x_2} = e^{x_1 + x_2} = f(x_1 + x_2)
\end{align}

## 로지스틱함수

**로지스틱(logistic)함수**는 지수함수를 변형한 함수로 시그모이드(sigmoid)함수의 하나다.

하지만 일반적으로 시그모이드함수라고 하면 이 로지스틱 함수를 뜻한다.

(시그모이드 함수 = 0과 1사이로 출력되는 함수)

\begin{align}
\sigma(x) = \dfrac{1}{1 + \exp(-x)} 
\end{align}

1. 지수함수는 승수가 음수여도 결국엔 양수가된다.
2. 따라서 1 + 지수함수 출력값은 1 이상이 된다.
3. 1 이상인 어떤 수의 역수이므로 1보다 작은 값을 가지게 된다.
4. 지수함수의 특징에 따라 0이상의 값을 가지게 된다.
5. 시그모이드 함수의 특징인 0과 1사이의 값을 갖는 함수가 된다.

```python
def logistic(x):
    return 1 / (1 + np.exp(-x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, logistic(xx))
plt.title('로지스틱함수')
plt.xlabel('$x$')
plt.ylabel('$\sigma(x)$')
plt.grid()
plt.show()
```

만약 지수함수에서 밑이 $e$가 아니라면 다음처럼 로그를 이용해 계산할 수 있다.

$y = a^x = (e^{log a})^x = e^{xloga}$

이처럼 로그는 지수함수의 출력이 특정한 값이 되게 하는 입력을 찾는 지수함수의 역함수이다.

$y = logx$

**로그함수의 특징**

2. $x>1 이면 y > 0 $ 양수
3. $x=1 이면 y=0$
4. $0<x<1 이면 y<0$음수
5. $x_1 > x_2 이면 logx_1 > logx_2$ (단조증가)

```python
xx = np.linspace(0.01, 8, 100)
yy = np.log(xx)
plt.title("로그함수")
plt.plot(xx, yy)
plt.axhline(0, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.axvline(1, c='r', ls="--")
plt.xlabel("$x$")
plt.ylabel("$\log(x)$")
plt.grid()
plt.show()
```

## 로그함수의 성질 1 : 로그함수는 곱하기를 더하기로 변환한다.

\begin{align}
\log{(x_1 \cdot x_2)} = \log{x_1} + \log{x_2} 
\end{align}

이 성질을 확장하면

\begin{align}
\log{\left(\prod_i x_i\right)} = \sum_i \left(\log{x_i}\right) 
\end{align}

위 식에서 $x_1 = x_2 = ... = x_n$이면

\begin{align}
\log x^n = n \log x 
\end{align}

## 로그함수의 성질 2 : 어떤 함수에 로그를 적용해도 함수의 최고점, 최저점의 위치는 변하지 않는다.

최적화를 하기 위해선 어떤 함수 $f$가 어떤 위치에서 최고, 최저점을 갖는지를 알아야한다.

그런데 어떤 함수의 경우 최고, 최저점을 알기 굉장히 어려울 수 있다.

따라서 이 함수에 로그를 취하는 것으로 새로운 함수를 만들 수 있는데 이렇게 새로운 함수의 최고, 최저점은 구하기 쉬워진다. 또한 로그를 취한 새로운 함수의 최고, 최저점은 원래 함수의 최고, 최저점과 동일하다.

\begin{align}
\arg\max_x f(x) = \arg\max_x \log f(x)
\end{align}

```python
def ff(x):
    return x**3 - 12*x + 20 * np.sin(x) + 7

xx = np.linspace(-4, 4, 300)
yy = ff(xx)

plt.subplot(211)
plt.plot(xx, yy)
plt.axhline(1, c='r', ls="--")
plt.yticks([0, 1, 5, 10])
plt.ylim(-2, 15)
plt.title("$f(x)$")
plt.grid()
plt.subplot(212)
plt.plot(xx, np.log(yy))
plt.axhline(0, c='r', ls="--")
plt.title("$log f(x)$")
plt.grid()
plt.tight_layout()
plt.show()
```

로그함수는 0부터 1사이의 구간을 음의 무한대부터 0까지 확장시켜 보여준다. 이러한 특징을 이용해 확률값처럼 0과 1사이의 값을 더 확실하게 비교할 수 있도록 해준다.

## 다변수함수

다변수함수란 **복수의 입력변수**를 갖는 함수이다.

예를 들어 다음 수식은 함수 $f$가 2개의 입력변수 $x, y$를 받아 변수 $z$를 출력하는 다변수 함수라는 뜻이다. 이러한 모양의 함수를 2차원 함수라고도 한다.

$z = f(x, y)$

위와 같은 함수의 예시로는 $x$위도, $y$경도를 입력받아 해발고도를 출력하는 함수가 있을 수 있다.

이러한 2차원 함수는 평면상의 지형과 유사하기 때문에 3차원 서피스플롯(surface plot), 컨투어플롯(contour plot)으로 나타낼 수 있다.

그 이상의 차원을 갖는 함수는 정상적인 방법으로는 시각화 하기 어렵다.

`예제`

$f(x, y) = 2x^2 + 6xy + 7y^2 - 26x - 54y +107$

```python
def f(x, y):
    return 2 * x**2 + 6 * x * y + 7 * y**2 - 26 * x - 54 * y + 107

xx = np.linspace(-3, 7, 100)
yy = np.linspace(-3, 7, 100)
X, Y = np.meshgrid(xx, yy)
Z = f(X, Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0.1)
ax.view_init(40, -110)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')
plt.title("서피스 플롯(Surface Plot)")
plt.show()
```

```python
CS = plt.contour(X, Y, Z, levels=np.logspace(0, 3, 10))
plt.clabel(CS, fmt="%d")
plt.title('컨투어 플롯')
plt.grid()
plt.show()
```

## 분리 가능 다변수함수

다변수함수 중에는 **단변수함수의 곱으로 표현** 가능한 다변수함수가 존재한다.

$f(x, y) = f_1(x)f_2(y)$

위와 같은 함수를 **분리 가능 다변수함수**라고 한다.

위에서 살펴본 서피스플롯, 컨투어플롯을 통해 2차원 함수는 3차원 공간상의 지형에 비유할 수 있는 것을 확인했다.

이러한 2차원 함수에서 $x$값을 어떠한 상수 $x_0$로 고정하거나

$x = x_0$

또는 $y$값을 어떤 상수 $y_0$로 고정하면

$y = y_0$

조정 가능한 변수가 하나가 되므로 2차원 함수를 1차원 단변수함수로 변형시킬 수 있다.

이 때 **단변수함수의 모양은 3차원 공간상의 지형을 케이크를 자르듯이 나눌 때 생기는 단면의 모양과 같아진다.**

여기서 중요한 특징은 분리 가능 다변수함수는 단면의 모양이 모두 같다는 것이다.

예를 들어 $x = x_0$로 고정한다면 단면함수는 다음과 같아진다.

$g(x_0, y) = g(x_0)g(y) = k_0g(y)$

따라서 $x_0$의 값과 상관없이 $g(y)$의 높이만 조절한 모양이 된다.

`예제`

\begin{align}
g(x, y) = \exp(-x^2 - 16y^2) = \exp(-x^2)\exp(-16y^2) = g_1(x)g_2(y)
\end{align}

위 함수에서 $y$의 값을 0, 0.1, 0.2, 0.3으로 고정하고 시각화해보자.

(두 지수함수의 곱은 입력값 합의 지수함수와 같다는 것을 기억하자.)

```python
from matplotlib import transforms
from matplotlib.ticker import NullFormatter

def g1(x):
    return np.exp(-x ** 2)

def g2(y):
    return np.exp(-16 * y **2)

def g(x, y):
    return g1(x) * g2(y)

xx = np.linspace(-1, 1, 100)
yy = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(xx, yy)
Z = g(X, Y)

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02
rect = [left, bottom, width, height]
rectx = [left, bottom_h, width, 0.2]
recty = [left_h, bottom, 0.2, height]

plt.figure(1, figsize=(8, 8))
ax = plt.axes(rect)
axx = plt.axes(rectx)
axy = plt.axes(recty)

nullfmt = NullFormatter()
axx.xaxis.set_major_formatter(nullfmt)
axy.yaxis.set_major_formatter(nullfmt)

ax.contour(X, Y, Z)
ax.axhline(0, c='r', ls="-")
ax.axhline(0.1, c='g', ls="--")
ax.axhline(0.2, c='b', ls="-.")
ax.axhline(0.3, c='c', ls=":")
ax.set_xlabel("x")
ax.set_ylabel("y")

axx.plot(xx, g1(xx), c='r')
axx.plot(xx, g2(0.1) * g1(xx), c='g', ls="--")
axx.plot(xx, g2(0.2) * g1(xx), c='b', ls="-.")
axx.plot(xx, g2(0.3) * g1(xx), c='c', ls=":")
axx.set_title("g1(x)")
axx.text(-0.2, 0.3, "y=0.3로 자른 단면")
axx.text(-0.2, 0.6, "y=0.2으로 자른 단면")
axx.text(-0.2, 0.9, "y=0.1로 자른 단면")

base = axy.transData
rot = transforms.Affine2D().rotate_deg(-90)
axy.plot(yy, g2(yy), transform=rot + base)
axy.set_title("g2(y)")
axy.axhline(0, xmax=g2(0), c='r', ls="-")
plt.text(0.05, 0.02, "$g_2$(0)={:.2f}".format(g2(0)))
axy.axhline(0.1, xmax=g2(0.1), c='g', ls="--")
plt.text(0.05, 0.12, "$g_2$(0.1)={:.2f}".format(g2(0.1)))
axy.axhline(0.2, xmax=g2(0.2), c='b', ls="-.")
plt.text(0.05, 0.22, "$g_2$(0.2)={:.2f}".format(g2(0.2)))
axy.axhline(0.3, xmax=g2(0.3), c='c', ls=":")
plt.text(0.05, 0.32, "$g_2$(0.3)={:.2f}".format(g2(0.3)))

axx.set_xlim(ax.get_xlim())
axx.set_ylim(0, 1)
axy.set_ylim(ax.get_ylim())
axy.set_xlim(0, 1)

plt.suptitle("$g(x, y) = \exp(-x^2)\exp(-16y^2) = g_1(x)g_2(y)$의 그래프", y=1.04)


plt.show()
```

## 다변수 다출력 함수

다변수 다출력 함수는 입력변수뿐 아니라 출력변수 또한 다수인 함수이다. 이 경우 출력값은 벡터 혹은 행렬로 나타낼 수 있다.

이후 딥러닝에서 다룰 **소프트맥스(softmax)**함수는 다차원 벡터를 입력값으로 받아 다차원 벡터를 출력한다.

아래는 3차원 소프트맥스 함수의 예시이다.

\begin{split} 
\begin{align}
y =
\begin{bmatrix}
y_1 \\ y_2 \\ y_3
\end{bmatrix}
=S(x_1, x_2, x_3) = 
\begin{bmatrix}
\dfrac{\exp(w_1x_1)}{\exp(w_1x_1) + \exp(w_2x_2) + \exp(w_3x_3)} \\
\dfrac{\exp(w_2x_2)}{\exp(w_1x_1) + \exp(w_2x_2) + \exp(w_3x_3)} \\
\dfrac{\exp(w_3x_3)}{\exp(w_1x_1) + \exp(w_2x_2) + \exp(w_3x_3)} \\
\end{bmatrix}
\end{align}
\end{split}

위 함수를 잘 살펴보면 각 벡터의 원소는 동일한 분모를 가지는 것을 알 수 있고 분자는 순서대로 들어가는 것을 확인할 수 있다.

또한 입력과 출력 사이의 연산 과정에서 벡터의 각 원소를 만들기 위해 모든 입력값을 사용하는 것을 확인할 수 있다.

위 함수의 연산과정을 살펴보면 다음의 특징을 찾을 수 있다.

1. 분자가 분모의 일부와 동일하다는 것을 통해 분모값이 더 큰것을 확인할 수 있다.
2. 1번 특징을 통해 위 함수의 출력은 0과 1사이의 값임을 알 수 있다.
3. 분모가 모두 동일하다는 특징을 통해 벡터의 각 원소합이 1이 됨을 알 수 있다.
4. 입력원소의 크기 순서와 출력 원소의 크기 순서가 같다.

위 특징들을 통해 소프트맥스함수는 **다변수 입력을 확률처럼 보이게 출력**하는 특징이 있다.

이러한 함수의 특성으로 인해 인공신경망의 마지막단에 출력을 조건부확률로 변형하는데 사용된다.

```python
def sofmax(x, w): # x는 1차원 배열, w는 가중치
    e = np.exp(w * x)
    return np.exp(w * x) / e.sum()

x = [2.0, 1.0, 0.5]
y = sofmax(x, np.ones(3))
y
```

```python
np.sum(y)
```

가중치를 크게 설정하면 최댓값과 최속값의 차이가 더 벌어진다.

```python
y = sofmax(x, 4 * np.ones(3))
y
```

## 함수의 평행이동

단변함수를 그래프 상에서 오른쪽으로 $a$만큼 평행이동하려면 다음과 같이 함수식을 변경한다.

$f(x) \rightarrow f(x-a)$

즉 함수식에서 $x$로 표기된 부분을 모두 $x-a$로 바꾸는 것이다.

예를 들어 함수

$f(x) = x^2 + 2x$

를 오른쪽으로 2만큼 이동하려면

$f(x-2) = (x-2)^2 + 2(x-2)$

로 변경하면 된다.

**빼기는 오른쪽 이동이고 더하기는 왼쪽이동이라는 점에 유의한다.**

단변수함수를 그래프 상에서 위로 $B$만큼 평행이동하려면 다음과 같이 함수식을 변경한다.

$f(x) \rightarrow f(x) + b$

**위로 이동은 더하고 아래로 이동은 뺀다.**

다음은 로지스틱함수를 오른쪽으로 5, 아래로 1만큼 평행이동한 예시이다.

```python
xx = np.linspace(-10, 10, 100)
plt.plot(xx, logistic(xx), label='$\sigma(x)$', ls='-')
plt.plot(xx, logistic(xx-5)-1, label='$\sigma(x-5)-1$', ls='--')
plt.legend()
plt.title('로지스틱함수를 오른쪽으로 5, 아래로 1만큼 이동')
plt.grid()
plt.show()
```

로지스틱함수는 0와 1사이에서 그려지며 $x$가 0일 때 가장 많은 변화가 일어난다.

반면에 아래로 5, 아래로 1 평행이동한 로지스틱 함수는 -1과 0 사이에서 그려지며 오른쪽으로 5만큼 이동했기에 가장 많은 변화가 일어나는 구간은 $x$가 5일 때다.

만약 변수가 둘인 다변수 함수를 오른쪽으로 $a$만큼, 위로$b$만큼 평행이동하려면 함수를 다음과 같이 변형한다.

$f(x, y) = f(x-a, y-b)$

```python
def g(x, y):
    return np.exp(-x**2 - 16 * y**2)

xx = np.linspace(-1, 1, 100)
yy = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(xx, yy)
Z1 = g(X, Y)
Z2 = g(X-0.5, Y-0.75)
plt.contour(X, Y, Z1)
plt.contour(X, Y, Z2, linestyles="--")
plt.text(-0.05, -0.02, "f(x,y)")
plt.text(0.35, 0.73, "f(x-0.5,y-0.75)")
plt.ylim(-0.5, 1)
plt.title("다변수함수의 평행이동")
plt.grid()
plt.show()
```

## 함수의 스케일링

단변수함수를 $x$축 방향으로 $a$배만큼 늘리려면 함수를 다음과 같이 변형한다.

\begin{align}
f(x) \;\; \rightarrow \;\; f\left(\frac{x}{a}\right)
\end{align}

단변수함수를 $y$축 방향으로 $b$만큼 늘리려면 함수를 다음과 같이 변형한다.

\begin{align}
f(x) \;\; \rightarrow \;\; bf(x)
\end{align}

아래는 로지스틱함수를 $x$축 방향으로 1/2배, $y$축 방향으로 2배한 것이다.

```python
xx = np.linspace(-6, 6, 100)
plt.plot(xx, logistic(xx), label="$\sigma(x)$", ls="-")
plt.plot(xx, 2*logistic(2*xx), label="2$\sigma(2x)$", ls="--")
plt.legend()
plt.title("로지스틱함수를 x축 방향으로 1/2배, y축 방향으로 2배")
plt.grid()
plt.show()
```

변형된 로지스틱함수는 $x$축 방향으로 2배 압축되었고 $y$축 방향으로 2배 증가하였다.

따라서 함수의 경사가 4배 급해졌다고 할 수 있으며 이러한 함수를 하이퍼탄젠트 함수라고 한다.

이는 딥러닝에서 로지스틱함수를 이용할 때 성능을 높이기 위해 사용한다.
