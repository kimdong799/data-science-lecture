### ipython 설정
- startup 파일 설정
    - 주피터 노트북 실행 시 미리 실행되는 코드 작성
    - 설정 방법
        - 프로필 파일 생성 : `$ ipython profile create`
        - startup 파일 수정
            - `$ cd ~/.ipython/profile_defauly/startup`
            - `$ vi 00-first.py`

- inline backend config
    - `c.IPKernelApp.matplotlib = 'inline'`
    - `c.InlineBackend.figure_format = 'retina`
