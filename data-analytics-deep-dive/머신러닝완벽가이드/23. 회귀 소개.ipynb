# 회귀(regression)란?
---

* 회귀를 설명하기 위해 먼저 위키백과에 정의된 회귀에 대한 설명을 살펴보겠습니다.

회귀(영어: regress 리그레스[*])의 원래 의미는 옛날 상태로 돌아가는 것을 의미한다. 영국의 유전학자 프랜시스 골턴은 부모의 키와 아이들의 키 사이의 연관 관계를 연구하면서 부모와 자녀의 키사이에는 선형적인 관계가 있고 키가 커지거나 작아지는 것보다는 전체 키 `평균으로 돌아가려는 경향`이 있다는 가설을 세웠으며 이를 분석하는 방법을 "회귀분석"이라고 하였다. 이러한 경험적 연구 이후, 칼 피어슨은 아버지와 아들의 키를 조사한 결과를 바탕으로 함수 관계를 도출하여 회귀분석 이론을 수학적으로 정립하였다.

출처 : https://ko.wikipedia.org/wiki/%ED%9A%8C%EA%B7%80_%EB%B6%84%EC%84%9D

위의 설명과 같이 회귀는 데이터 값이 평균과 같은 일정한 값으로 돌아가려는 경향을 이용한 통계학 기법으로 여러 분야의 발전에 크게 기여했습니다.

위의 회귀에 대한 설명을 통계학적으로 설명하자면 회귀란 여러 독립변수와 하나의 종속변수간의 상관관계를 모델리하는 기법을 말합니다.

# 머신러닝의 회귀
---

머신러닝은 크게 지도 학습과 비지도 학습으로 나뉘어지는데 지도 학습은 다시 연속형 변수를 예측하는 회귀와 이산형 변수를 예측하는 분류로 나뉘어집니다.







**그렇다면 머신러닝에서의 회귀는 무엇일까?**

머신러닝의 (선형)회귀는 실제 값과 예측값의 차이를 최소화하는 식을 찾는 것으로, y=wx+b의 식에서 독립변수의 값에 영향을 미치는 회귀 계수(Regression Coefficients)의 최적의 값을 찾는 것입니다.





# 선형 회귀의 유형
---

회귀는 한번 더 독립변수의 개수와 회귀 계수의 선형/비선형 여부로 크게 4가지로 구분됩니다.

**(단일회귀, 다중회귀, 선형회귀, 비선형회귀)**

위의 회귀 중 가장 많이 사용되는 것은 선형 회귀이며 과적합 문제를 해경하기 위한 회귀 계수에 규제를 적용하는 방법에 따라 아래와 같이 구분할 수 있습니다.

* `일반 선형 회귀` : 예측값과 실제값의 RSS(Residual Sum of Squares)를 최소화 하도록 회귀계수를 최적화, 규제를 적용하지 않음


* `릿지(Ridge)회귀` : 선형회귀에 L2 규제를 추가, L2 규제는 상대적으로 큰 회귀 계수 값의 예측 영향도를 감소시키기 위해 회귀 계수값을 더 작게 만드는 모델


* `라쏘(Lasso)회귀` : 선형회귀에 L1 규제를 추가, L1 규제는 예측 영향력이 작은 피처의 회귀 계수를 0으로 만들어 예측 시 피처가 선택되지 않게 하는 모델(L1 규제는 피처 선택 기능으로도 불림)


* `엘라스틱넷(ElasticNet)` : L2, L1 규제를 함께 결합한 모델. 주로 피처가 많은 데이터 세트에 적용, L1 규제로 피처를 줄이고 L2규제로 계수 값의 크기를 조정


* `로지스틱 회귀(Logstic Regression)` : 회귀라는 이름이 쓰이지만 분류에 사용되는 선형 모델. 이진분류, 희소 영역 분류(텍스트 분류) 등에 뛰어난 예측 성능을 보임

# 단순 선형 회귀
---

단순 선형 회귀란 무엇일까요?

**단순 선형 회귀란 독립변수(X)와 종속변수(y)가 하나 씩 존재하는 회귀**를 의미합니다. 

예를 들어 부모님의 키에 따라 자녀의 키가 결정된다고 가정할 때 부모의 키가 클수록 유전적인 영향에 의해 자녀의 키가 커지는 경향을 확인할 수 있는데 이는 **선형 관계**로 표현할 수 있습니다.

이미지 참고 : https://ko.wikipedia.org/wiki/%EC%84%A0%ED%98%95_%ED%9A%8C%EA%B7%80

이처럼 선형회귀를 통해 위와 같이 Data points들을 대표하는 Linear regression을 생성하여 새로운 X값에 대해 y값을 예측하는 것이 가능해집니다. 

여기서 예측값은 **y^(y 햇)으로 표현하며 y^ = wx + b**로 계산이 가능해집니다. 

또한 **실제 Data points들은 선형식과의 잔차가 존재할 수 있어 wx + b + Error(잔차) 혹은 wx + b - Error(잔차)** 로 계산됩니다.

정리하면 선형 회귀를 통해 해당 데이터 포인트를 대표할 수 있는 선형식을 계산한 다음 새로운 데이터에 대해 선형식을 적용하여 y^을 예측하는 것 입니다.

# 최적의 회귀 모델을 생성하는 방법
---

최적의 회귀 모델을 생성한다는 말의 의미는 **잔차의 합이 최소가 되는 선형식을 도출**하는 것을 의미합니다. 다시 말해 앞서 설명한 **최적의 회귀 계수를 찾는 것**이라고 봐도 무방합니다.

이미지 출처: https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=hancury&logNo=220173717142

위 도표에서 알 수 있듯이 잔차는 +, - 값이 모두 가능하기 때문에 잔차를 단순히 더하게 되면 0이 되는 경우가 발생할 수 있어 일반적으로 절댓값을 취하거나(**MAE, Mean Absolute Error**) 제곱 합의 평균을 구하여 사용합니다.(**RSS, Residual Sum of Square**)

여기서 RSS에 대한 내용은 선형회귀의 핵심으로 더 자세히 알아봅시다.

위에서 머신러닝을 통한 선형회귀의 목적은 잔차를 최소로 가지는 회귀 계수를 도출하는 것이라 설명했습니다.  여기서 말하는 잔차는 **RSS가 될 것이며 즉, RSS를 최소로 가지는 회귀계수 w0, w1을 기계 학습을 통해 도출**하는 것입니다.

이렇게 RSS와 같이 머신러닝 모델이 계속해서 학습할지를 결정하는 것을 **비용함수 혹은 손실함수(Loss Function)**이라고 합니다.

먼저 **예측값 y^은 wx + b**로 계산된다는 것을 기억합시다.

그리고 **오차의 제곱 합의 평균(MSE)는 손실함수**로 아래와 같은 수식을 사용합니다.

# 경사하강법 구현
---

위에서 경사하강법에 사용되는 gdW와 gdB에 대해 다루었습니다.

이번에는 위에서 도출한 식을 파이썬 코드로 구현해 실제 경사하강법이 어떻게 진행되는지 시각적으로 표현하는 내용을 다루어보겠습니다. (위 링크에서 테디노트님의 강의를 참고하시면 좋습니다.)

**gdW : LR * ((y^ - y ) * x)MEAN**

**gdB : LR * ((y^ - y ))MEAN**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 표준정규분포를 따르는 0 ~ 1 사이의 난수 발생
X = np.random.rand(100)

# 임의의 노이즈를 위해 아래와 같이 y변수 생성
y = X + np.random.rand(100)
```

```python
# 위에서 생성한 X, y 시각화

plt.scatter(X, y)
plt.show()
```

```python
# 실제값과 예측값을 시각화 하는 함수

def plot_prediction(pred, y):
    plt.scatter(X, y, label = 'real_value')
    plt.scatter(X, pred, label = 'prediction')
    plt.legend()
    plt.show()
```

```python
# Gradient Desent 구현

# 초기 w, b 값
w = np.random.uniform(-1, 1)
b = np.random.uniform(-1, 1)

LR = 0.5

for epoch in range(100):
    y_Pred = w * X + b # 예측값
    
    # error가 0.001 이하이면 early stopping 하도록
    error = np.abs(y_Pred - y).mean() # MAE
    if error < 0.001:
        break 
        
    # Gradient Descent 계산
    w_grad = LR * ((y_Pred - y) * X).mean()
    b_grad = LR * ((y_Pred - y)).mean()
    
    # 새로운 w, b 갱신
    w = w - w_grad
    b = b - b_grad
    
    if epoch % 20 == 0:
        y_Pred = w * X + b
        plot_prediction(y_Pred, y)
```

위 결과를 통해 실제 경사하강법을 통해 어떻게 예측값이 실제 값에 가까워지는지 시각적으로 확인해보았습니다.
